# Python: Fundamentos y Herramientas de Cómputo

Python es un lenguaje de programación de alto nivel creado por **Guido van Rossum** a inicios de los noventa. Su filosofía se centra firmemente en la legibilidad del código, lo que permite a los desarrolladores expresar conceptos en pocas líneas en comparación con lenguajes como C++ o Java.

### Pilares Fundamentales:

- **Paradigma Orientado a Objetos:** En Python, prácticamente **todo es un objeto** (desde un número entero hasta una función o un módulo). Esto significa que cada elemento posee atributos (datos) y métodos (funciones integradas).
- **Tipado Dinámico:** Las variables no están amarradas a un tipo de dato estático. El intérprete deduce el tipo en tiempo de ejecución, permitiendo reasignar una variable libremente.
- **Naturaleza Interpretada:** No requiere compilación previa. El código se procesa línea por línea, lo cual encaja perfectamente con el flujo de trabajo interactivo de Jupyter Notebooks.

| Ventajas | Desventajas |
| :--- | :--- |
| Sintaxis limpia, intuitiva y cercana al lenguaje humano | Ejecución más lenta en tareas puramente secuenciales que C |
| Ecosistema gigante para Ciencia de Datos y Estadística | Mayor consumo de memoria por la gestión dinámica de objetos |

## Operadores en Python

Los operadores aritméticos en Python se utilizan para realizar operaciones matemáticas básicas. A continuación se presentan los más comunes, con ejemplos y explicaciones detalladas.

### 1. Suma
Suma dos números. Si se usan con cadenas, las concatena, pero en este contexto nos centramos en numéricos.

```python
# Suma de enteros
s = 3 + 2
print(s) 

# Suma de flotantes
s2 = 2.5 + 1.5
print(s2)  

# Mezcla de tipos: el resultado es flotante
s3 = 3 + 2.0
print(s3) 
```

### 2. Resta
Resta el segundo número del primero.

```python
r = 4 - 7
print(r)  # -3

# Con flotantes
r2 = 10.5 - 3.2
print(r2)  # 7.3
```

### 3. Negación (`-` como operador unitario)
Cambia el signo de un número.

```python
n = -7
print(n)  # -7

# También puede aplicarse a variables
x = 5
y = -x
print(y)  # -5
```

### 4. Multiplicación (`*`)
Multiplica dos números.

```python
m = 2 * 6
print(m)  # 12

# Con flotantes
m2 = 2.5 * 4
print(m2)  # 10.0
```

### 5. Exponente (`**`)
Eleva el primer número a la potencia del segundo.

```python
e1 = 2 ** 6
print(e1)  # 64

# Exponentes con flotantes
e2 = 9 ** 0.5  # raíz cuadrada
print(e2)  # 3.0

# Exponentes negativos
e3 = 2 ** -1
print(e3)  # 0.5
```

### 6. División (`/`)
Realiza la división y siempre devuelve un número flotante (decimal), incluso si los operandos son enteros.

```python
d = 3.5 / 2
print(d)  # 1.75

# División de enteros produce flotante
d2 = 7 / 2
print(d2)  # 3.5

# División entre cero genera error
# d3 = 5 / 0  # ZeroDivisionError
```

### 7. División entera (`//`)
Devuelve la parte entera del cociente (truncamiento hacia abajo, no redondeo). Es decir, el resultado es el mayor entero menor o igual al cociente exacto.

```python
de = 3.5 // 2
print(de)  # 1.0  (porque 3.5/2 = 1.75, parte entera 1)

# Con números negativos, el truncamiento es hacia abajo
de2 = -3.5 // 2
print(de2)  # -2.0  (porque -3.5/2 = -1.75, el mayor entero menor o igual es -2)

# División entera entre enteros da entero
de3 = 7 // 2
print(de3)  # 3
```

### 8. Operador módulo (`%`)
Devuelve el resto de la división entre dos números. Es muy útil para saber si un número es par o múltiplo de otro.

```python
mod = 7 % 2
print(mod)  # 1  (porque 7 dividido 2 da 3, y resto 1)

# Otros ejemplos
print(10 % 3)  # 1
print(15 % 5)  # 0  (múltiplo)
print(8 % 3.5) # 1.0 (con flotantes)

# Con números negativos, el signo del resto sigue al divisor
print(-7 % 2)  # 1  (en Python, el resto siempre es no negativo)
```

#### Consideraciones adicionales

- **Prioridad de operadores**: Siguen las reglas matemáticas habituales (potencia antes que multiplicación/división, y estas antes que suma/resta). Se pueden usar paréntesis para cambiar el orden.
- **Tipos de datos**: Al mezclar enteros y flotantes, el resultado suele ser flotante.
- **Errores comunes**: División por cero, o esperar un entero en división normal.

Los valores booleanos también resultan de comparaciones entre valores.

| Operador | Descripción                  | Ejemplo             | Resultado |   
|----------|------------------------------|---------------------|-----------|
| `==`     | ¿Son iguales a y b?          | `r = 5 == 3`        | `False`   |
| `!=`     | ¿Son distintos a y b?        | `r = 5 != 3`        | `True`    |
| `<`      | ¿Es a menor que b?           | `r = 5 < 3`         | `False`   |
| `>`      | ¿Es a mayor que b?           | `r = 5 > 3`         | `True`    |
| `<=`     | Menor o igual que            | `5 <= 5`            | `True`    |
| `>=`     | Mayor o igual que            | `5 >= 3`            | `True`    |

## Listas en Python

Las **listas** son una de las estructuras de datos más versátiles y utilizadas en Python. Nos permiten almacenar una colección de elementos (que incluso pueden ser de diferentes tipos, como números y texto) en una sola variable.

**Características principales:**
* Se definen utilizando corchetes `[ ]` y separando sus elementos con comas.
* Tienen un orden específico (podemos acceder a sus elementos por su posición).
* Son mutables (podemos modificar sus elementos después de crearla).

**Comportamiento con operadores matemáticos:**
A diferencia de los vectores matemáticos, los operadores en las listas nativas de Python tienen un comportamiento enfocado en la estructura, no en el cálculo aritmético:
* El operador `*` **repite** los elementos de la lista.
* El operador `+` sirve para **concatenar** (unir) dos listas, por lo que intentar sumar un número entero directamente a una lista nos arrojará un error de tipo (`TypeError`).

````python
# Creación y alteración de listas
lenguajes = ["R", "SQL", "C++"]
lenguajes.insert(0, "Python")  # Añade al inicio del índice
lenguajes.pop()                # Remueve el último elemento ("C++")
print("Lista resultante:", lenguajes)

# Slicing avanzado sobre una secuencia numérica
serie_numerica = list(range(20))
print("Sublista (índices del 5 al 11):", serie_numerica[5:12])
print("Inversión total de la lista:", serie_numerica[::-1])
````

### Numpy (Numerical Python) - Vectores y Matrices

Las listas de Python no están diseñadas para operaciones matemáticas directas elemento a elemento. Para ello usamos **NumPy**, que introduce el concepto de **vectorización**: ejecutar cálculos matemáticos masivos en bloques contiguos de memoria de manera asombrosamente veloz.

NumPy es la biblioteca fundamental para la computación científica en Python. Proporciona un objeto de arreglo multidimensional de alto rendimiento y herramientas para trabajar con estos arreglos. Si alguna vez has trabajado con MATLAB, encontrarás muchas similitudes.

El corazón de NumPy es el arreglo (array) , una estructura de datos que almacena elementos del mismo tipo (generalmente números) de forma contigua en memoria. Esto permite realizar operaciones matemáticas extremadamente rápidas, ya que las operaciones se ejecutan en C en lugar de en bucles de Python.

### Creación de arreglos básicos

````python
import numpy as np

# Creación de arreglos matriciales mediante distribución aleatoria y rangos
matriz_identidad_mod = np.eye(3) * 5       # Matriz con 5s en la diagonal principal
datos_normales = np.random.normal(loc=0, scale=1, size=(3, 3)) # Distribución Gaussiana

print("Matriz Diagonal:\n", matriz_identidad_mod)
print("\nMatriz Gaussiana 3x3:\n", datos_normales)

# Multiplicación de matrices usando el operador moderno '@'
matriz_A = np.array([[2, 4], [1, 3]])
matriz_B = np.array([[5, 0], [2, 1]])
producto_matricial = matriz_A @ matriz_B
print("\nResultado del Producto Matricial:\n", producto_matricial)
````

````python
# Crear un arreglo unidimensional (vector) a partir de una lista
vector = np.array([1, 2, 3, 4, 5])
print("Vector:", vector)

# Crear un arreglo bidimensional (matriz)
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Matriz:\n", matriz)

# Funciones útiles para crear arreglos
ceros = np.zeros((3, 4))        # Matriz de 3x4 llena de ceros
unos = np.ones((2, 3))           # Matriz de 2x3 llena de unos
rango = np.arange(0, 11, 2)      # Similar a range() pero devuelve un arreglo: [0, 2, 4, 6, 8, 10] (no llega hasta 11)
lineal = np.linspace(0, 1, 5)    # 5 puntos espaciados uniformemente entre 0 y 1: [0., 0.25, 0.5, 0.75, 1.]
identidad = np.eye(3)             # Matriz identidad de 3x3

print("\nCeros:\n", ceros)
print("Unos:\n", unos)
print("Rango:", rango)
print("Lineal:", lineal)
print("Identidad:\n", identidad)
````

### Operaciones matemáticas básicas

Una de las mayores ventajas de NumPy es que permite realizar operaciones vectorizadas. En lugar de escribir bucles, las operaciones se aplican elemento por elemento automáticamente.

#### Operaciones aritméticas elemento a elemento

````python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Operaciones aritméticas
print("Suma:", a + b)          # [11 22 33 44]
print("Resta:", a - b)          # [9 18 27 36]
print("Multiplicación:", a * b)  # [10 40 90 160]
print("División:", a / b)        # [10. 10. 10. 10.]
print("Potencia:", a ** 2)       # [100 400 900 1600]

# También con escalares
print("Multiplicar por escalar:", a * 2)  # [20 40 60 80]
````

#### Operaciones de agregación (suma, producto, etc.)

````python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Matriz original:\n", matriz)

print("Suma total:", np.sum(matriz))           # 21
print("Suma por columnas (eje 0):", np.sum(matriz, axis=0))  # [5 7 9]
print("Suma por filas (eje 1):", np.sum(matriz, axis=1))     # [6 15]

print("Media total:", np.mean(matriz))          # 3.5
print("Mínimo:", np.min(matriz))                # 1
print("Máximo:", np.max(matriz))                # 6
print("Desviación estándar:", np.std(matriz))   # 1.7078...
````

Además, NumPy incluye un submódulo llamado `linalg` para operaciones de álgebra lineal y también ofrece funciones estadísticas rápidas y eficientes (mean, std).

````python
import pandas as pd

# Vectores numéricos independientes
id_alumnos = np.array([101, 102, 103, 104, 105])
horas_estudio = np.array([12, 18, 5, 22, 15])
calificacion_base = np.array([75, 88, 60, 95, 70])

# Cálculo vectorizado: Ajustamos la calificación según horas de estudio
calificacion_final = calificacion_base + (horas_estudio * 0.2)
calificacion_final = np.clip(calificacion_final, 0, 100) # Forzamos a que el límite máximo sea 100

# Construcción del DataFrame a partir de un diccionario
df_estudiantes = pd.DataFrame({
    'ID': id_alumnos,
    'Horas_Estudio': horas_estudio,
    'Nota_Base': calificacion_base,
    'Nota_Final': calificacion_final
})

df_estudiantes
````

### Segmentación Dinámica y Filtros Booleanos

Para interrogar al DataFrame bajo múltiples condiciones, usamos operadores bitwise (`&` para AND, `|` para OR) y obligatoriamente aislamos cada bloque con paréntesis.

````python
# Filtro 1: Estudiantes que dedicaron más de 14 horas al estudio
alto_esfuerzo = df_estudiantes[df_estudiantes['Horas_Estudio'] > 14]
print("Alumnos con Alto Esfuerzo ")
display(alto_esfuerzo)  #

# Filtro 2: Alumnos con una Nota_Base menor a 80 PERO que alcanzaron una Nota_Final mayor o igual a 75
filtro_mejora = df_estudiantes[(df_estudiantes['Nota_Base'] < 80) & (df_estudiantes['Nota_Final'] >= 75)]
print("\nAlumnos que repuntaron en la Nota Final")
display(filtro_mejora)  

# Resumen estadístico global del grupo
print("\n Sumario Estadístico Descriptivo")
display(df_estudiantes.describe().T) #  La tabla del resumen estadístico transpuesta
````
