## 1. Ley de los Grandes Números (LGN)

### Contexto Histórico
* **Formuló:** Jacob Bernoulli (1730) — *Ley Débil para variables independientes Bernoulli.*
* **Formalización:** Andrei Kolmogorov (1933) — *Estableció las bases axiomáticas que permitieron la Ley Fuerte.*

> "Establece que el promedio muestral converge (en algún sentido) al valor esperado conforme el tamaño de la muestra aumenta."

### Concepto 
Si repites un experimento aleatorio muchas veces, la media de los resultados observados se aproximará cada vez más al valor esperado teórico (la media poblacional $\mu$). En términos simples: **el azar a la larga se estabiliza**.

### Definición 
Sea $X_1, X_2, \dots, X_n$ una secuencia de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) con $E[X_i] = \\mu$ y $Var(X_i) = \sigma^2 < \infty$.

* **Ley Débil:** El promedio muestral $\bar{X}_n$ converge en probabilidad a $\mu$:
$$\lim_{n \to \infty} P(|\bar{X}_n - \mu| > \varepsilon) = 0 \quad \forall \varepsilon > 0$$

* **Ley Fuerte:** El promedio muestral converge casi seguramente (con probabilidad 1) a $\mu$:
$$P\left(\lim_{n \\to \infty} \bar{X}_n = \mu\right) = 1$$

#### Comportamiento de la Suma ($S_n$)
Definimos la variable de acumulación o suma como:
$$S_n = \sum_{i=1}^{n} X_i = X_1 + X_2 + \dots + X_n$$

* **Esperanza de la Suma:** Por la propiedad de linealidad del operador esperanza, la esperanza de la suma es la suma de las esperanzas individuales:
$$\mathbb{E}[S_n] = \mathbb{E}\left[\sum_{i=1}^{n} X_i\right] = \sum_{i=1}^{n} \mathbb{E}[X_i] = \sum_{i=1}^{n} \mu = n\mu$$

* **Varianza de la Suma:** Debido a la condición de independencia mutua, todas las covarianzas cruzadas se anulan ($\text{Cov}(X_i, X_j) = 0$ para $i \neq j$), lo que simplifica la varianza como:
$$\text{Var}(S_n) = \text{Var}\left[\sum_{i=1}^{n} X_i\right] = \sum_{i=1}^{n} \text{Var}(X_i) = \sum_{i=1}^{n} \sigma^2 = n\sigma^2$$

#### Comportamiento del Promedio Muestral ($\bar{X}_n$)
El promedio se define escalando la suma por el tamaño muestral $n$:
$$\bar{X}_n = \frac{S_n}{n} = \frac{1}{n}\sum_{i=1}^{n} X_i$$

* **Esperanza del Promedio:** El promedio muestral actúa como un estimador insesgado:
$$\mathbb{E}[\bar{X}_n] = \mathbb{E}\left[\frac{S_n}{n}\right] = \frac{1}{n}\mathbb{E}[S_n] = \frac{1}{n}(n\mu) = \mu$$

* **Varianza del Promedio:** Al extraer la constante del operador varianza, esta se eleva al cuadrado:
$$\text{Var}(\bar{X}_n) = \text{Var}\left[\frac{S_n}{n}\right] = \frac{1}{n^2}\text{Var}(S_n) = \frac{1}{n^2}(n\sigma^2) = \frac{\sigma^2}{n}$$

> **Conclusión de la LGN:** Cuando $n \to \infty$, la varianza del promedio muestral se desvanece ($\frac{\sigma^2}{n} \to 0$). El azar se estabiliza y la media muestral converge al valor esperado teórico $\mu$.

### Definición Formal de Convergencia
* **Ley Débil:** El promedio muestral $\bar{X}_n$ converge en probabilidad a $\mu$:
$$\lim_{n \to \infty} P(|\bar{X}_n - \mu| > \varepsilon) = 0 \quad \forall \varepsilon > 0$$

* **Ley Fuerte:** El promedio muestral converge casi seguramente (con probabilidad 1) a $\mu$:
$$P\left(\lim_{n \to \infty} \bar{X}_n = \mu\right) = 1$$

```python
# Importamos las librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Fijamos una semilla para reproducibilidad
np.random.seed(42)
```

```python
# Vamos a simular 10,000 lanzamientos de un dado
# np.random.randint(low, high, size) genera enteros desde 'low' (inclusive) hasta 'high' (exclusive)
lanzamientos = np.random.randint(1, 7, size=10000)# entre 1 y 6
```

```python
# Calculamos la suma acumulada de los lanzamientos
suma_acom = np.cumsum(lanzamientos)

# Creamos el número de experimentos realizados en cada paso (1 hasta 10,000)
n = np.arange(1, 10001)

# Obtenemos el promedio muestral en cada paso
promedios_muestrales = suma_acom / n
```

```python
#creamos el data frame 

df = pd.DataFrame({
    'Lanzamiento': np.arange(1, 10001),
    'Media Acomulada': promedios_muestrales})
df
```

```python
# creamos la gráfica
plt.figure(figsize=(10,6))

# Usamos los nombres exactos de tus columnas: 'Lanzamiento' y 'Media Acomulada'
plt.plot(df['Lanzamiento'], df['Media Acomulada'], label='media muestral', color='blue')

# Usamos plt.axhline con 'y=' para trazar la línea horizontal en 3.5
plt.axhline(y=3.5, color='red', linestyle='--', label='media teórica')

plt.xlabel('Número de lanzamientos')
plt.ylabel('Media acomulada')
plt.title('Simulación de la LGN de lanzamiento de un dado')
plt.legend()
plt.grid(True, linestyle=':') # Opcional: una cuadrícula ayuda a ver mejor la convergencia
plt.show()
```

# El problema de Monty Hall

Es un famoso dilema de probabilidad basado en un concurso de televisión, y es un gran ejemplo de cómo las intuiciones humanas pueden ser engañosas en contextos probabilísticos.

## Planteamiento del problema:

Tienes tres puertas enfrente de ti. Detrás de una de ellas hay un coche (el premio), y detrás de las otras dos hay cabras (premios no deseados).

Tu elección inicial: Eliges una puerta, pero no la abres aún.

Monty Hall (el presentador, que sabe qué hay detrás de cada puerta) abre una de las otras dos puertas, revelando una cabra.

Segunda oportunidad: Ahora Monty te da la opción de cambiar de puerta (de tu elección original a la otra puerta que queda cerrada) o quedarte con tu elección inicial.

### ¿Qué deberías hacer? ¿Cambiar de puerta o quedarte con tu elección original?

* **Respuesta correcta:** Deberías cambiar de puerta.
* **Explicación:** Inicialmente, al elegir una puerta, tienes una probabilidad de $1/3$ de haber elegido el coche y una probabilidad de $2/3$ de haber elegido una cabra.

Cuando Monty abre una puerta revelando una cabra, no cambia estas probabilidades. Al principio tenías un $2/3$ de probabilidades de haber elegido mal, y Monty, al mostrar una cabra, te está ayudando a confirmar que el coche probablemente está detrás de la otra puerta.

Si elegiste mal inicialmente (lo cual sucede en el $66.67\%$ de los casos), Monty te da la oportunidad de corregir tu elección cambiando a la puerta correcta.

Si elegiste correctamente desde el principio (probabilidad del $33.33\%$), entonces cambiar haría que perdieras.

Por lo tanto, cambiar de puerta te da una probabilidad de ganar de $2/3$, mientras que quedarte con tu elección original te deja con solo una probabilidad de ganar de $1/3$.

### Ejemplo:
* **Puertas:** $[1, 2, 3]$
* El coche está detrás de la puerta 2.
* Si eliges la puerta 1 (probabilidad $1/3$), Monty abre la puerta 3, que tiene una cabra.
* Si cambias a la puerta 2, ganas el coche (probabilidad $2/3$).

## 2. Teorema del Límite Central (TLC)

### Contexto Histórico y Evolución
El Teorema del Límite Central se utilizó por primera vez con el objetivo de **aproximar una distribución Binomial a una distribución Normal**. Su desarrollo formal clave incluye:

* **Abraham de Moivre (1733):** Descubrió el primer caso particular del teorema al aproximar la distribución de la suma de lanzamientos de monedas (Binomial con $p=0.5$) mediante la curva normal (*Teorema de De Moivre-Laplace*).
* **Pierre-Simon de Laplace (1812):** Popularizó y extendió el teorema, generalizándolo para variables Binomiales con cualquier probabilidad de éxito $p$.
* **Aleksandr Lyapunov (1901):** Consiguió una demostración matemática rigurosa y general del teorema para variables independientes que no necesariamente tienen la misma distribución, introduciendo la poderosa herramienta de la **función característica**.

Mientras que la LGN te dice *a dónde* converge el promedio (a la media $\mu$), el TLC te dice *cómo* se distribuye ese promedio en el camino. No importa la distribución original de tus datos (puede ser exponencial, uniforme, binomial, etc.), **la suma o el promedio de una muestra grande de variables aleatorias tenderá a distribuirse como una Normal**.

### Definición Formal
Sea $X_1, X_2, \dots, X_n$ una secuencia de variables aleatorias i.i.d. con $E[X_i] = \mu$ y $Var(X_i) = \sigma^2 < \infty$. Conforme $n \to \infty$, la variable estandarizada $Z_n$ converge en distribución a una Normal estándar $\mathcal{N}(0,1)$:

$$Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}} \xrightarrow{d} \mathcal{N}(0,1)$$

O de forma equivalente, para el promedio muestral:
$$\bar{X}_n \sim \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)$$

### Caso Especial: Aproximación de la Binomial a la Normal (De Moivre-Laplace)


### Ejemplo Práctico: Aproximación por TLC

**Enunciado:** Lanzamos una moneda justa y sea $X$ el número de soles ("caras"). Realizamos el experimento $n = 5000$ veces.
$$X \sim \text{Bin}(5000, 1/2)$$

Queremos calcular la probabilidad de obtener entre 3500 y 4500 soles:
$$\mathbb{P}(3500 \le X \le 4500)$$

#### El problema analítico (Cálculo Directo)
Intentar resolver esto mediante la función de masa de probabilidad Binomial requeriría calcular una sumatoria masiva:
$$\mathbb{P}(3500 \le X \le 4500) = \sum_{i=3500}^{4500} \binom{5000}{i} \left(\frac{1}{2}\right)^i \left(1 - \frac{1}{2}\right)^{5000-i}$$
Calcular términos como $\binom{5000}{3500}$ es computacional y analíticamente inviable a mano. Por lo tanto, recurrimos a aproximar usando el TLC.

#### Solución mediante Aproximación Normal
1. **Calcular los parámetros teóricos:**
   * **Valor Esperado ($\mathbb{E}[X]$):** $$\mathbb{E}[X] = n \cdot p = 5000 \cdot \frac{1}{2} = 2500$$
   * **Varianza ($\text{Var}(X)$):** $$\text{Var}(X) = n \cdot p \cdot (1-p) = 5000 \cdot \frac{1}{2} \cdot \frac{1}{2} = 1250$$
   * **Desviación Estándar ($\sigma$):** $$\sigma = \sqrt{\text{Var}(X)} = \sqrt{1250} \approx 35.3553$$

2. **Estandarizar la variable ($Z$):**
   De acuerdo al TLC, podemos transformar nuestra variable a una distribución normal estándar $\mathcal{N}(0,1)$ mediante:
   $$Z = \frac{X - \mathbb{E}[X]}{\sqrt{\text{Var}(X)}} \sim \mathcal{N}(0,1)$$

3. **Planteamiento de la probabilidad estandarizada:**
   $$\mathbb{P}(3500 \le X \le 4500) = \mathbb{P}\left( \frac{3500 - 2500}{\sqrt{1250}} \le \frac{X - 2500}{\sqrt{1250}} \le \frac{4500 - 2500}{\sqrt{1250}} \right)$$
   $$\mathbb{P}(3500 \le X \le 4500) = \mathbb{P}\left( \frac{1000}{35.3553} \le Z \le \frac{2000}{35.3553} \right)$$
   $$\mathbb{P}(3500 \le X \le 4500) \approx \mathbb{P}(28.28 \le Z \le 56.56)$$

*(Nota teórica: Dado que los valores z son extremadamente altos, esta probabilidad es prácticamente 0, lo cual tiene sentido lógico ya que esperar obtener más de 3500 soles en 5000 lanzamientos de una moneda justa es un evento extremadamente raro, alejado por completo de la media teórica de 2500).*

```python
# CÓDIGO PYTHON: Simulación del TLC 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)
num_simulaciones = 10000
tamaños_n = [1, 2, 5, 30]  

fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=False)
axes = axes.flatten()

mu_teorica = 0.5
var_teorica = 1/12

for idx, n in enumerate(tamaños_n):
    muestras = np.random.uniform(0, 1, size=(num_simulaciones, n))
    promedios = np.mean(muestras, axis=1)
    
    # Histograma interactivo/limpio: Azul cielo suave sin bordes negros rígidos
    axes[idx].hist(promedios, bins=50, density=True, alpha=0.7, color='#9ecae1')
    
    # Curva normal teórica en azul marino obscuro profundo
    sigma_n = np.sqrt(var_teorica) / np.sqrt(n)
    x = np.linspace(mu_teorica - 4*sigma_n, mu_teorica + 4*sigma_n, 200)
    axes[idx].plot(x, stats.norm.pdf(x, mu_teorica, sigma_n), color='navy', lw=2.5)
    
    axes[idx].set_title(f'Tamaño de muestra n = {n}', fontsize=11, pad=8)
    axes[idx].set_xlabel('X', fontsize=10)
    axes[idx].set_ylabel('Densidad', fontsize=10)
    
    # Limpieza de recuadro (Spines) y agregado de rejilla sutil
    axes[idx].spines['top'].set_visible(False)
    axes[idx].spines['right'].set_visible(False)
    axes[idx].spines['left'].set_color('#cccccc')
    axes[idx].spines['bottom'].set_color('#cccccc')
    axes[idx].grid(axis='y', linestyle='-', alpha=0.2, color='#999999')

plt.suptitle('Teorema del Límite Central: De una Uniforme a una Normal', fontsize=13, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```

```python
# CÓDIGO PYTHON: Simulación del TLC mostrando el efecto de n - Estilo Limpio
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)
num_simulaciones = 10000
tamaños_n = [1, 2, 5, 30]  # Diferentes tamaños de muestra para ver la evolución

fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=False)
axes = axes.flatten()

# Datos de origen: Uniforme(0,1) -> Media = 0.5, Varianza = 1/12
mu_teorica = 0.5
var_teorica = 1/12

for idx, n in enumerate(tamaños_n):
    # Generamos muestras de tamaño 'n' para cada una de las 10,000 simulaciones
    muestras = np.random.uniform(0, 1, size=(num_simulaciones, n))
    promedios = np.mean(muestras, axis=1)
    
    # Histograma con el azul claro/celeste suave característico (sin bordes blancos marcados)
    axes[idx].hist(promedios, bins=50, density=True, alpha=0.7, color='#9ecae1')
    
    # Curva normal teórica correspondiente en azul marino profundo (navy)
    sigma_n = np.sqrt(var_teorica) / np.sqrt(n)
    x = np.linspace(mu_teorica - 4*sigma_n, mu_teorica + 4*sigma_n, 200)
    axes[idx].plot(x, stats.norm.pdf(x, mu_teorica, sigma_n), color='navy', lw=2.5)
    
    # Títulos y etiquetas por subgráfico
    axes[idx].set_title(f'Tamaño de muestra n = {n}', fontsize=11, pad=8)
    axes[idx].set_xlabel('X', fontsize=10)
    axes[idx].set_ylabel('Densidad', fontsize=10)
    
    # Ajustes estéticos (Limpieza de recuadro y rejilla sutil horizontal)
    axes[idx].spines['top'].set_visible(False)
    axes[idx].spines['right'].set_visible(False)
    axes[idx].spines['left'].set_color('#cccccc')
    axes[idx].spines['bottom'].set_color('#cccccc')
    axes[idx].grid(axis='y', linestyle='-', alpha=0.2, color='#999999')

plt.suptitle('Teorema del Límite Central: De una Uniforme a una Normal', fontsize=13, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```

Cuando modelamos $S_n \sim \text{Binomial}(n, p)$, si $n$ es suficientemente grande ($np \ge 5$ y $n(1-p) \ge 5$), la distribución de los éxitos se comporta de forma aproximada como:
$$S_n \sim \mathcal{N}\big(np, np(1-p)\big)$$

Formalmente, la variable estandarizada $Z_n$ se define y distribuye como:
$$Z_n = \frac{S_n - n\mu}{\sqrt{n\sigma^2}} = \frac{S_n - n\mu}{\sqrt{n}\sigma} \sim \mathcal{N}(0,1)$$

> **Nota de convergencia:** Conforme $n \to \infty$, $Z_n$ **converge a una normal** estándar $\mathcal{N}(0,1)$.

###  Función de la Densidad de la Media Muestral
Indica que, conforme el tamaño de la muestra tiende a infinito ($n \to \infty$), la distribución de la media muestral $\bar{X}_n$ se aproxima a una distribución Normal, sin importar la distribución original de las variables independientes e idénticamente distribuidas (v.a. i.i.d.).

Formalmente, la función de densidad de la media muestral $\bar{X}_n$ converge en distribución a una densidad Normal con los siguientes parámetros:
$$\bar{X}_n \xrightarrow{d} \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)$$

Donde su función de densidad de probabilidad continua viene dada de forma asintótica por:
$$f_{\bar{X}_n}(x) \approx \frac{1}{\frac{\sigma}{\sqrt{n}}\sqrt{2\pi}} \exp\left( -\frac{1}{2}\left(\frac{x - \mu}{\sigma / \sqrt{n}}\right)^2 \right)$$

> **Nota de convergencia:** Al estandarizar esta variable mediante la transformación $Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}}$, su densidad **converge a una normal** estándar $\mathcal{N}(0,1)$.

## 3. De Variables Aleatorias a Vectores Aleatorios

### A. Variable Aleatoria Unidimensional
Una variable aleatoria $X$ es una función que asigna un número real a cada resultado posible del espacio muestral $\Omega$.

$$X: \Omega \to \mathbb{R}$$

* Para un evento elemental $\omega_i \in \Omega$, su imagen es $x_i = X(\omega_i) \in \mathbb{R}$.
* A partir de este mapeo, podemos definir la **Función de Distribución Acumulada (FDA)** en la recta real como:
$$F_X(x) = \mathbb{P}(X \le x)$$

### B. Vector Aleatorio Bidimensional (Caso Bivariado)
Cuando nos interesa estudiar dos características simultáneas de un mismo fenómeno aleatorio sobre el mismo espacio muestral $\Omega$, extendemos el concepto a un par ordenado de funciones $(X, Y)$.

$$(X, Y): \Omega \to \mathbb{R}^2$$

* Para cada evento elemental $\omega_i \in \Omega$, el vector asigna un punto en el plano cartesiano:
$$(X(\omega_i), Y(\omega_i)) = (x, y) \in \mathbb{R}^2$$

#### Condición de Existencia
El par ordenado $(X, Y)$ es formalmente un **vector aleatorio (bivariado)** si y solo si sus componentes individuales son variables aleatorias bajo el mismo espacio de probabilidad. Es decir:
$$\text{Si } X \text{ es v.a. } \quad \text{y} \quad Y \text{ es v.a. } \implies (X, Y) \text{ es un vector aleatorio.}$$

#### Función de Distribución Acumulada Conjunta
La probabilidad de que el vector aleatorio tome valores en una región semi-infinita del plano (abajo y a la izquierda del punto $(x,y)$) se define mediante la **FDA conjunta**:

$$F_{(X,Y)}(x,y) = \mathbb{P}(X \le x, Y \le y)$$

Geométricamente, esto representa la probabilidad acumulada en el tercer cuadrante si desplazáramos temporalmente el origen de coordenadas al punto $(x, y)$.

## 4. Generalización y Distribuciones Conjuntas Discretas

### A. Vector Aleatorio de Dimensión $n$
Sean $X_1, X_2, \dots, X_n$ variables aleatorias definidas sobre el mismo espacio de probabilidad. Un **vector aleatorio de dimensión $n$** es un mapeo o función que va desde el espacio muestral $\Omega$ hacia el espacio euclídeo de $n$ dimensiones $\mathbb{R}^n$:

$$\mathbf{X}: \Omega \to \mathbb{R}^n$$

Donde expresamos el vector como el par ordenado de sus componentes:
$$\mathbf{X} = (X_1, X_2, \dots, X_n)$$

### B. Función de Masa de Probabilidad Conjunta (Caso Discreto)
Para un vector aleatorio bidimensional discreto $(X, Y)$, la distribución conjunta de probabilidades se describe mediante su **Función de Masa de Probabilidad (FMP) conjunta**, denotada por $f_{(X,Y)}(x,y)$, la cual asigna la probabilidad formal de que ocurra la intersección exacta de los eventos $\{X = x\}$ y $\{Y = y\}$:

$$f_{(X,Y)}(x,y) = \begin{cases} \mathbb{P}(X = x, Y = y), & \text{si } (x,y) \in \mathcal{R}_{(X,Y)} \\ 0, & \text{e.o.c.} \end{cases}$$

Donde:
* $\mathcal{R}_{(X,Y)}$ representa el soporte conjunto o el conjunto de todos los valores posibles que el vector aleatorio puede tomar.
* `e.o.c.` significa *en otro caso*.

*(Nota didáctica: A menudo en las notas de clase o discusiones informales nos referimos de forma general a este concepto como **densidad conjunta**, aunque formalmente el término "densidad" se reserva para variables continuas e integrales, y "masa" para variables discretas y sumatorias).*

### Propiedades Fundamentales de la FMP Conjunta
Para que una función $f_{(X,Y)}(x,y)$ sea una función de masa de probabilidad conjunta legítima, debe satisfacer obligatoriamente las siguientes dos propiedades:

1. **No Negatividad:** La probabilidad de cualquier par de valores no puede ser negativa.
   $$f_{(X,Y)}(x,y) \ge 0 \quad \forall (x,y) \in \mathbb{R}^2$$

2. **Condición de Normalización (Suma Unitaria):** La suma de las probabilidades sobre todo el soporte conjunto debe ser exactamente igual a 1 (certeza absoluta sobre el espacio muestral).
   $$\sum_{x} \sum_{y} f_{(X,Y)}(x,y) = 1$$

## 5. Distribuciones Conjuntas Continuas

### Función de Densidad de Probabilidad Conjunta
Si $(X, Y)$ es un **vector aleatorio continuo**, su comportamiento probabilístico conjunto ya no se define a través de sumatorias, sino mediante integrales en el plano real. Su función de densidad conjunta $f_{(X,Y)}(x,y)$ debe cumplir rigurosamente con dos propiedades análogas:

1. **No Negatividad:** La función de densidad conjunta es no negativa en todo su dominio.
   $$f_{(X,Y)}(x,y) \ge 0 \quad \forall (x,y) \in \mathbb{R}^2$$

2. **Condición de Normalización (Integral Unitaria):** El volumen total bajo la superficie de la función de densidad sobre todo el espacio bidimensional debe ser exactamente igual a 1.
   $$\int_{\mathbb{R}} \int_{\mathbb{R}} f_{(X,Y)}(x,y) \, dx dy = 1$$

## 6. Distribuciones Marginales

A partir de la función de distribución conjunta, es posible aislar el comportamiento probabilístico de una sola de las variables componentes del vector eliminando (o colapsando) la influencia de la otra variable. Estas funciones resultantes se conocen como **Funciones Marginales**.

### A. Caso Discreto (Funciones de Masa Marginales)
Para un vector aleatorio bidimensional discreto, las funciones de probabilidad individuales se obtienen sumando la función conjunta sobre todos los valores posibles de la variable que se desea omitir:

* **Distribución Marginal de $X$ ($f_X(x)$):** Se colapsa el eje de las $Y$ sumando verticalmente todas las probabilidades conjuntas asociadas a un valor fijo de $x$:
$$f_X(x) = \sum_{y} f_{(X,Y)}(x,y)$$

* **Distribución Marginal de $Y$ ($f_Y(y)$):** Se colapsa el eje de las $X$ sumando horizontalmente todas las probabilidades conjuntas asociadas a un valor fijo de $y$:
$$f_Y(y) = \sum_{x} f_{(X,Y)}(x,y)$$

### B. Caso Continuo (Funciones de Densidad Marginales)
Para un vector aleatorio continuo $(X, Y)$, las funciones de densidad individuales individuales se obtienen integrando la función conjunta sobre todo el soporte de la otra variable ($\mathbb{R}$):

* **Función de Densidad Marginal de $X$ ($f_X(x)$):** Se integra respecto a $y$. Analíticamente equivale a acumular todo el volumen de la "montaña" a lo largo de las franjas de $Y$:
$$f_X(x) = \int_{\mathbb{R}} f_{(X,Y)}(x,y) \, dy$$

* **Función de Densidad Marginal de $Y$ ($f_Y(y)$):** Se integra respecto a $x$. Analíticamente equivale a acumular todo el volumen de la superficie a lo largo de las franjas de $X$:
$$f_Y(y) = \int_{\mathbb{R}} f_{(X,Y)}(x,y) \, dx$$

## 7. Independencia de Variables Aleatorias

###  Independencia de Eventos 
Antes de definir la independencia en variables aleatorias, se establece la base axiomática a partir de los eventos dentro del espacio muestral. Decimos que dos eventos $A$ y $B$ son independientes si la probabilidad de que ocurran ambos simultáneamente (su intersección) es igual al producto de sus probabilidades individuales:

$$\mathbb{P}(A \cap B) = \mathbb{P}(A)\mathbb{P}(B)$$

### Concepto Clave
Decimos que dos variables aleatorias $X$ e $Y$ son **independientes** si el conocimiento sobre la ocurrencia o valor de una de ellas no altera en absoluto la distribución de probabilidad de la otra. En términos operativos, la distribución conjunta puede factorizarse exactamente como el producto de sus respectivas distribuciones marginales.


###  Independencia de Variables Aleatorias
Extendiendo la definición anterior, dos variables aleatorias $X$ e $Y$ son **independientes** si la ocurrencia o el valor de una de ellas no afecta de ninguna manera la distribución de probabilidad de la otra. En términos prácticos, su distribución conjunta se puede factorizar por completo como el producto de sus marginales.

#### Caso Discreto
Para dos variables aleatorias discretas $X$ e $Y$, la condición clave de independencia se cumple si y solo si para cualquier par de valores $(x, y)$ se verifica que:
$$\mathbb{P}(X = x, Y = y) = \mathbb{P}(X = x)\mathbb{P}(Y = y)$$

O expresado mediante sus funciones de masa de probabilidad conjunta y marginales:
$$f_{(X,Y)}(x,y) = f_X(x) \cdot f_Y(y)$$

#### Caso Continuo
Para un vector aleatorio continuo $(X, Y)$, la noción se mantiene idéntica pero aplicada a sus densidades. $X$ e $Y$ son independientes si y solo si la función de densidad conjunta es igual al producto de sus densidades marginales para todo punto del plano $(x, y) \in \mathbb{R}^2$:
$$f_{(X,Y)}(x,y) = f_X(x) \cdot f_Y(y)$$

###  Generalización para $n$ Variables Aleatorias
En general, el concepto puede extenderse a una secuencia de $n$ variables aleatorias $X_1, X_2, \dots, X_n$. Decimos que estas variables son mutuamente independientes si su función de distribución conjunta (ya sea de masa o de densidad) es numéricamente igual al producto de cada una de sus funciones marginales individuales:

$$f_{\mathbf{X}}(x_1, \dots, x_n) = f_{X_1}(x_1) \cdot \dots \cdot f_{X_n}(x_n)$$

Utilizando la notación matemática compacta de productorias, la condición formal se define como:
$$f_{\mathbf{X}}(x_1, \dots, x_n) = \prod_{i=1}^{n} f_{X_i}(x_i)$$

###  Variables Aleatorias Independientes e Idénticamente Distribuidas (v.a. i.i.d.)
Si además de cumplir con la condición de independencia mutua, las variables aleatorias $X_1, X_2, \dots, X_n$ tienen la misma distribución de probabilidad, se concluye que comparten de manera idéntica sus momentos estadísticos principales:

1. **Mismo Valor Esperado:** Todas poseen el mismo centro de masa o media poblacional ($\mu$).
   $$\mathbb{E}[X_i] = \mu \quad \forall i$$

2. **Misma Varianza:** Todas poseen el mismo nivel de dispersión o varianza poblacional ($\sigma^2$).
   $$\text{Var}(X_i) = \sigma^2 \quad \forall i$$

Cuando las variables son independientes y comparten la misma distribución de probabilidad (v.a. i.i.d.), los momentos poblacionales se unifican de forma simétrica para todo el vector:
* **Vector de Medias:** $\mathbb{E}[\mathbf{X}] = (\mu, \mu, \dots, \mu)^T$
* **Matriz de Covarianzas ($\Sigma$):** Se vuelve una matriz diagonal pura dada la ausencia de correlación ($\text{Cov}(X_i, X_j) = 0$):
$$\Sigma = \sigma^2 I_n$$
