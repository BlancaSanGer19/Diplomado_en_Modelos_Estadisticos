# 1. Conceptos Fundamentales de Probabilidad

---

### Definición de Espacio de Probabilidad $(\Omega, \mathcal{F}, \mathbb{P})$

Establecido por **Andrei Kolmogorov en 1933**, este modelo unifica la probabilidad clásica con el rigor de la Teoría de la Medida (Borel y Lebesgue). Un espacio de probabilidad es una terna matemática compuesta por:

1. **$\Omega$:** Un conjunto no vacío que representa el **espacio de estados** (la totalidad de los resultados posibles del experimento aleatorio).
2. **$\mathcal{F}$:** Una familia de subconjuntos de $\Omega$ que cumple con tres condiciones estrictas:
   * Contiene al conjunto vacío: $\emptyset \in \mathcal{F}$.
   * Cerrada bajo complementos: Si $A \in \mathcal{F} \implies A^c \in \mathcal{F}$.
   * Cerrada bajo uniones numerables: Si $A_1, A_2, \dots \in \mathcal{F} \implies \bigcup_{n=1}^{\infty} A_n \in \mathcal{F}$.
3. **$\mathbb{P}$:** Una medida de probabilidad, definida como una función $\mathbb{P}: \mathcal{F} \rightarrow [0,1]$. que asigna una cuantificación numérica a la incertidumbre.


#### Axiomas de Kolmogorov (1933):
* **Axioma 1 (No negatividad):** Para cualquier evento $A \in \mathcal{F}$, $0 \le \mathbb{P}(A)$.
* **Axioma 2 (Normalización):** La probabilidad del espacio total (evento seguro) es la certeza absoluta: $\mathbb{P}(\Omega) = 1$.
* **Axioma 3 ($\sigma$-aditividad):** Si $(A_n)_{n \ge 1}$ es una sucesión de eventos mutuamente excluyentes o disjuntos dos a dos ($A_i \cap A_j = \emptyset$ para $i \ne j$), entonces la probabilidad de su unión numerable es la suma de sus probabilidades individuales:
$$\mathbb{P}\left(\bigcup_{n=1}^{\infty} A_n\right) = \sum_{n=1}^{\infty} \mathbb{P}(A_n)$$

#### Propiedades Derivadas Básicas de la Probabilidad:
* **Evento Imposible:** $\mathbb{P}(\emptyset) = 0$
* **Complemento (Evento Contrario):** $\mathbb{P}(A^c) = 1 - \mathbb{P}(A)$
* **Monotonía / Diferencia de Conjuntos:** Si $A \subseteq B \implies \mathbb{P}(B \setminus A) = \mathbb{P}(B) - \mathbb{P}(A)$. De esto se deriva que si $A \subseteq B \implies \mathbb{P}(A) \le \mathbb{P}(B)$.
* **Principio de Inclusión-Exclusión (Regla de Adición):** $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B) - \mathbb{P}(A \cap B)$


> **Interpretación:** La terna de Kolmogorov no define qué *es* la probabilidad en el mundo real (si es una frecuencia o un grado de creencia), sino cómo debe *comportarse* matemáticamente para evitar paradojas en el infinito continuo. La $\sigma$-álgebra $\mathcal{F}$ representa algebraicamente la estructura de la información disponible o los eventos que somos capaces de observar y "medir".
> 
> **Conclusión:** Los axiomas de Kolmogorov transformaron la probabilidad de un conjunto de reglas empíricas para juegos de azar en una rama formal de las matemáticas. Sin esta estructura, sería imposible fundamentar la teoría de procesos estocásticos y los modelos de cálculo actuarial de largo plazo.

### 1.1 Experimentos Aleatorios, Espacios Muestrales y Eventos

Un **experimento aleatorio ($E$)** es aquel fenómeno que, al reproducirse bajo condiciones idénticas, puede conducir a varios resultados posibles sin que se pueda predecir el desenlace de forma determinista antes de su ejecución.

#### Clasificación del Espacio Muestral ($\Omega$)
Es el conjunto de todos los resultados posibles. Dependiendo de la naturaleza de sus elementos ($\omega \in \Omega$), se clasifica en:

**1. Espacios Muestrales Discretos (Finitos o Infinitos Numerables):**
* *Finitos:* Lanzamiento de un dado honesto $\implies \Omega = \{1, 2, 3, 4, 5, 6\}$ (Cardinalidad $|\Omega| = 6$).
* *Infinitos Numerables:* Número de lanzamientos de una moneda hasta obtener la primera águila $\implies \Omega = \{1, 2, 3, \dots\} = \mathbb{N}$ (Cardinalidad $\aleph_0$).

**2. Espacios Muestrales Continuos (Infinitos No Numerables):**
* *Intervalos Acotados:* Tiempo de espera de un asegurado en una unidad de emergencia (en horas) $\implies \Omega = [0, 24]$.
* *Intervalos No Acotados:* Tiempo de vida de un componente electrónico o la duración de una póliza hasta el siniestro $\implies \Omega = [0, \infty)$.

Ejemplo:


* **Lanzamiento de una moneda:** $\Omega = \{\text{águila}, \text{sol}\}$

* **Lanzamiento de un dado:** $\Omega = \{1, 2, 3, 4, 5, 6\}$

* **Tiempo de espera de Romeo a Julieta (en horas):** $\Omega = [0, 1]$

* **Tiempo de vida de una bacteria:** $\Omega = [0, \infty)$

#### Eventos Aleatorios y Estructura de Conjuntos
Un evento $A$ es matemáticamente un elemento de la $\sigma$-álgebra $\mathcal{F}$ (un subconjunto medible de $\Omega$). 
* **Conjunto Potencia ($\mathbb{P}(\Omega)$):** En espacios muestrales finitos, la $\sigma$-álgebra natural adoptada es el conjunto potencia, el cual agrupa a absolutamente todos los subconjuntos posibles de resultados. Si $\Omega$ es finito, el número total de eventos evaluables es:
$$|\mathbb{P}(\Omega)| = 2^{|\Omega|}$$$ 



> **Interpretación:** Identificar correctamente la naturaleza geométrica o discreta de $\Omega$ dicta la herramienta analítica a utilizar. Si el espacio es finito o contable, la probabilidad se calcula mediante sumatorias y conteos combintatorios; si el espacio es continuo, la masa de probabilidad se distribuye de manera suave y su cuantificación requiere el uso de cálculo integral y medidas de longitud, área o volumen.
> 
> **Conclusión:** La correcta definición del experimento y de su espacio muestral constituye el paso inicial y crítico de cualquier modelación estadística. Un error en la delimitación del soporte $\Omega$ invalida los cálculos subsecuentes de la medida de probabilidad $\mathbb{P}$.

## 2.1 Probabilidad Condicional

La **probabilidad condicional** nos permite incorporar nueva información o "evidencia" al cálculo de probabilidades. En lugar de evaluar la probabilidad de que ocurra un evento $A$ sobre todo el espacio muestral original $\Omega$, calculamos la probabilidad de que ocurra $A$ **sabiendo** o dado que ya ha ocurrido (o es seguro que ocurrirá) un evento previo $B$.

Formalmente, si $\mathbb{P}(B) > 0$, la probabilidad condicional de $A$ dado $B$ se define como:

$$\mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)}$$

### Interpretación y Enfoque Clásico (Laplaciano)
Desde la perspectiva clásica, al saber que el evento $B$ ha ocurrido, nuestro espacio de resultados posibles se **reduce** o se "actualiza", pasando de ser $\Omega$ a ser únicamente el conjunto $B$. Los únicos casos que ahora son favorables para $A$ son aquellos resultados que están simultáneamente en $A$ y en $B$ (es decir, la intersección $A \cap B$). Por lo tanto, la fórmula se puede simplificar mediante las cardinalidades de los conjuntos:

$$\mathbb{P}(A \mid B) = \frac{\frac{|A \cap B|}{|\Omega|}}{\frac{|B|}{|\Omega|}} = \frac{|A \cap B|}{|B|}$$

### Propiedades Claves:
1. **Regla del Producto (Multiplicativa):** A partir de la definición, podemos despejar la probabilidad de la intersección. Esto es sumamente útil para calcular la ocurrencia simultánea de eventos secuenciales:
   $$\mathbb{P}(A \cap B) = \mathbb{P}(B) \cdot \mathbb{P}(A \mid B)$$
2. **Conexión con el Teorema de Bayes:** La probabilidad condicional es la piedra angular del análisis bayesiano, permitiéndonos invertir el orden de la condicionalidad mediante la relación:
   $$\mathbb{P}(B \mid A) = \frac{\mathbb{P}(A \mid B)\mathbb{P}(B)}{\mathbb{P}(A)}$$


> **Interpretación:** La probabilidad condicional opera bajo el concepto de **actualización de información**. En la modelación de riesgos, la ocurrencia del evento $B$ (por ejemplo, que un asegurado alcance la edad de 60 años) elimina de manera absoluta los escenarios complementarios del espacio muestral original $\Omega$, forzando una re-ponderación de las probabilidades de ocurrencia del evento $A$ (fallecer en el siguiente año) bajo el nuevo entorno de certeza $B$.
>
> **Conclusión:** La probabilidad condicional demuestra que la probabilidad no es una propiedad estática de un evento, sino una medida dinámica dependiente del estado de información disponible en el entorno del observador.

## 2.2 Independencia de Eventos

En la teoría de la probabilidad, la noción de independencia formaliza la idea de que la ocurrencia de un evento no aporta ninguna información sobre la ocurrencia del otro.

### Definición Formal
Decimos que dos eventos $A$ y $B$ son **independientes** si y solo si la probabilidad de su intersección es el producto de sus probabilidades individuales:

$$\mathbb{P}(A \cap B) = \mathbb{P}(A)\mathbb{P}(B)$$

### Forma Alternativa (Probabilidad Condicional)
Si $\mathbb{P}(B) > 0$, existe una forma alternativa y muy intuitiva de definir la independencia basándonos en el concepto de probabilidad condicional:

$$\mathbb{P}(A \mid B) = \mathbb{P}(A)$$

### Independencia vs. Mutua Exclusión
Existe una confusión muy común entre estos dos conceptos:
* **Eventos mutuamente excluyentes (o disjuntos):** Son eventos que **no pueden ocurrir al mismo tiempo** ($A \cap B = \emptyset$, por lo que $\mathbb{P}(A \cap B) = 0$). Si ocurre $A$, la probabilidad de que ocurra $B$ se vuelve automáticamente cero. Por lo tanto, ¡están fuertemente relacionados!
* **Eventos independientes:** Son eventos que **sí pueden ocurrir al mismo tiempo**, pero la presencia de uno no afecta al otro. 

*Regla de oro:* Si dos eventos tienen probabilidad mayor a cero, **no pueden ser mutuamente excluyentes e independientes al mismo tiempo**.


> **Interpretación:** Si conocer la ocurrencia de $B$ no modifica la medida de probabilidad de $A$, entonces la trayectoria del evento $B$ carece de poder predictivo sobre $A$. En esquemas de seguros de vida conjunta o riesgos correlacionados, asumir falsamente independencia matemática puede subestimar de forma severa la probabilidad de siniestros simultáneos.
>
> **Conclusión:** La independencia simplifica drásticamente el cálculo de probabilidades conjuntas a través de la factorización de sus medidas; sin embargo, representa una condición matemática restrictiva que debe validarse empíricamente antes de ser asumida en un modelo de riesgos.

## 2.3 Modelo Clásico (Laplaciano)

La **Probabilidad Clásica o Laplaciana** es un enfoque axiomático que se utiliza bajo dos condiciones fundamentales:
1. El espacio muestral $\Omega$ debe ser **finito**.
2. Todos los resultados elementales deben ser **equiprobables** (es decir, tienen la misma posibilidad de ocurrir, como en una moneda o un dado honesto).

Propuesto por Pierre-Simon Laplace (1812), se basa en el principio de **equiprobabilidad** (todos los eventos simples tienen la misma oportunidad de ocurrir).
Si se cumplen estas condiciones y sea $A$ un evento ($A \subseteq \Omega$), definimos la probabilidad de ocurrencia de $A$ como la proporción entre los casos favorables y los casos totales: 

$$\mathbb{P}(A) = \frac{\text{Casos Favorables}}{\text{Casos Totales}} = \frac{|A|}{|\Omega|}.$$

### Conceptos Relevantes
* **Axiomas de Kolmogorov:** $\mathbb{P}(A) \geq 0$, $\mathbb{P}(\Omega) = 1$, y para eventos excluyentes $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B)$.
* **Probabilidad del Complemento:** $\mathbb{P}(A^c) = 1 - \mathbb{P}(A)$.
* **Principio de Inclusión-Exclusión:** $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B) - \mathbb{P}(A \cap B)$.

### Formalización Estocástica (Distribución Uniforme Discreta)
El modelo laplaciano se modela formalmente a través de una variable aleatoria $X$ discreta que sigue una distribución uniforme sobre un soporte finito $\{x_1, x_2, \dots, x_k\}$, donde $|\Omega| = k$.

* **Función de Masa de Probabilidad (fmp):**
$$f(x) = \mathbb{P}(X = x) = \frac{1}{k}, \quad \text{para todo } x \in \Omega$$

* **Esperanza Matemática $\mathbb{E}[X]$:**
$$\mathbb{E}[X] = \sum_{i=1}^{k} x_i \cdot \mathbb{P}(X=x_i) = \frac{1}{k} \sum_{i=1}^{k} x_i$$

* **Varianza $\text{Var}(X)$:**
$$\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \left( \frac{1}{k} \sum_{i=1}^{k} x_i^2 \right) - \left( \mathbb{E}[X] \right)^2$$


> **Interpretación:** Bajo el esquema de Laplace, el analista asume un estado de ignorancia perfecta previa a la experimentación (Principio de Razón Insuficiente). Al no existir elementos que justifiquen ponderar un escenario por encima de otro, la masa de probabilidad se distribuye de manera estrictamente equitativa a lo largo del soporte.
>
> **Conclusión:** El enfoque laplaciano es puramente deductivo y lógico. Su utilidad radica en juegos de azar y muestreos probabilísticos ideales, pero resulta inoperante para el análisis de riesgos de fenómenos de la naturaleza o comportamiento humano donde la simetría no existe.

## 2.4 Enfoque Frecuentista

Este enfoque define la probabilidad a partir de la experimentación empírica repetida. Formulada originalmente por **Jacob Bernoulli**, la **Ley de los Grandes Números** establece que si realizamos un experimento independiente un número extremadamente grande de veces, la **frecuencia relativa** de un evento $A$ convergerá casi con total certeza a su **probabilidad teórica** $P(A)$.

### Formulación Matemática
La frecuencia relativa $f_n(A)$ se define como:
$$f_n(A) = \frac{n(A)}{n}$$

Donde $n(A)$ es el número de veces que el evento $A$ ocurre en las repeticiones y $n$ es el número total de ensayos.
El postulado frecuentista afirma que:
$$\lim_{n \to \infty} f_n(A) = \mathbb{P}(A)$$

### Formalización Estocástica (Ensayos de Bernoulli)
Cada ensayo dentro de la experimentación frecuentista se modela como una variable aleatoria de Bernoulli $X_i \sim \text{Bernoulli}(p)$, donde obtener el evento $A$ se denota con $1$ y su complemento con $0$.

* **Función de Masa de Probabilidad (fmp):**
$$f(x; p) = p^x (1-p)^{1-x}, \quad x \in \{0, 1\}$$

* **Esperanza y Varianza Marginal:**
$$\mathbb{E}[X_i] = p$$
$$\text{Var}(X_i) = p(1-p)$$

* **Comportamiento Asintótico (Ley Débil de los Grandes Números):**
Sea $\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$ el promedio muestral (frecuencia relativa), para cualquier $\epsilon > 0$:
$$\lim_{n \to \infty} \mathbb{P}(|\bar{X}_n - p| \geq \epsilon) = 0$$

* **Teorema Central del Límite (TCL):** Es el sustento matemático para la inferencia estadística; establece que la distribución de la suma (o promedio) de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) tiende a una **Distribución Normal** $\mathcal{N}(0, 1)$ conforme $n \to \infty$:
$$\frac{\bar{X}_n - \mathbb{E}[\bar{X}_n]}{\sqrt{\text{Var}(\bar{X}_n)}} = \frac{\bar{X}_n - p}{\sqrt{\frac{p(1-p)}{n}}} \xrightarrow{d} \mathcal{N}(0, 1)$$

#### Diferencia Conceptual
| Característica | Probabilidad Laplaciana | Probabilidad Frecuentista |
| :--- | :--- | :--- |
| **Enfoque** | Deductivo (Lógico) | Inductivo (Empírico) |
| **Origen** | Simetría del espacio muestral | Observación de frecuencias |
| **Condición** | Espacio finito y equiprobable | Experimentos repetibles (i.i.d.) |

> **Interpretación:** Para el enfoque frecuentista, la probabilidad no existe de forma abstracta a priori, sino que se descubre mediante la estabilización a largo plazo del comportamiento empírico. Es la base del diseño de tarifas de seguros: al observar el registro histórico de miles de siniestros independientes, la frecuencia observada colapsa hacia la verdadera tasa de riesgo implícita.
>
> **Conclusión:** Este enfoque permite transitar de los datos empíricos a la inferencia de parámetros poblacionales. Su limitación principal es que es inapropiado para eventos únicos, de baja frecuencia o no repetibles bajo las mismas condiciones.

## 2.5 Modelo Geométrico

La probabilidad geométrica extiende el concepto clásico de Laplace a **espacios muestrales infinitos y continuos** (no numerables), donde el conteo discreto de casos ya no es posible. En su lugar, la probabilidad se calcula como una proporción de magnitudes geométricas: **longitudes, áreas o volúmenes**.

### Formulación Matemática
Si $\Omega$ es un espacio muestral continuo que representa una región geométrica acotada en $\mathbb{R}^d$ y $A$ es un evento que corresponde a una subregión medible de $\Omega$, la probabilidad de que ocurra el evento $A$ se define mediante la medida de Lebesgue $\mu$:

$$\mathbb{P}(A) = \frac{\mu(A)}{\mu(\Omega)} = \frac{\text{Medida de la región deseada (Longitud / Área / Volumen)}}{\text{Medida de la región total}}$$

### Formalización Estocástica (Distribución Uniforme Continua)
Se asume que la variable aleatoria bidimensional o unidimensional distribuye de forma perfectamente uniforme sobre la región geométrica. Para el caso unidimensional en el intervalo $[a, b]$:

* **Función de Densidad de Probabilidad (fdp):**
$$f(x) = \begin{cases} \frac{1}{b - a} & \text{si } a \le x \le b \\ 0 & \text{en otro caso} \end{cases}$$

* **Esperanza Matemática $\mathbb{E}[X]$:**
$$\mathbb{E}[X] = \int_{a}^{b} x \cdot \frac{1}{b-a} \, dx = \frac{a + b}{2}$$

* **Varianza $\text{Var}(X)$:**
$$\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{(b - a)^2}{12}$$


> **Interpretación:** La probabilidad geométrica traslada la hipótesis de equiprobabilidad al continuo, estableciendo que la probabilidad de que un punto caiga dentro de una subregión específica es directamente proporcional a la dimensión (longitud, área) de dicha subregión, independientemente de su ubicación exacta dentro del espacio muestral.
>
> **Conclusión:** Este enfoque rompe las limitaciones del conteo elemental y da la bienvenida formal al cálculo integral para la medición matemática del azar. Es una herramienta potente para modelar la ocurrencia de eventos distribuidos de forma homogénea en el tiempo o el espacio continuo.

## Ejemplo 1: Problema del encuentro

Dos personas planean encontrarse en un lugar especifico, en un rango de una hora $[0,60]$ minutos (1 hora).
  - Si $A$ llega primero, entonces piensa esperar 10 minutos.
  - Si $B$ llega primero, entonces piensa esperar 10 minutos.

Se elige un número al azar entre $[0,60]$ y supongamos que el par de números se elige sobre el cuadro de modo que la probabilidad de un evento sea el área favorable entre el área total.

Sea: 
- $A$: las dos personas se encunetran a una distancia de 10 minutos
  
Nos gustaría encontrar la probabilidad $A$.


```python
import numpy as np
import matplotlib.pyplot as plt # para gráficos

# Vamos a generar 1000 situaciones
muestra = 20000

# Generamos los tiempos aleatorios entre [0,60]
x = np.random.uniform(0, 60, muestra) # numero aleatorio
y = np.random.uniform(0, 60, muestra)

# Añadimos una condición de encuentro
encuentro = np.where(abs(x-y) <= 10, 'r', 'b') # rojo si encuentran, azul si no.

# Ahora calculamos la probabilidad de que se encuentren
k = 0
for elemento in encuentro:
    if elemento == 'r': # Los favorables
        k += 1 # La añada
        
print("La probabilidad del evento A es: {}.".format(k/muestra))

# Graficamos
plt.figure(figsize=(5,5)) # Tamaño de la figura
plt.scatter(x, y, c=encuentro, s=1)
# Título principal
plt.title("Simulación del Problema del Encuentro", fontsize=14, fontweight='bold', color='#104E8B', pad=15)

# Títulos de ejes
plt.xlabel("Tiempo de llegada de la Persona X (minutos)", fontsize=11, color='#333333')
plt.ylabel("Tiempo de llegada de la Persona Y (minutos)", fontsize=11, color='#333333')

# Ajustamos los límites de los ejes exactos de 0 a 60 para que se vea limpio
plt.xlim(0, 60)
plt.ylim(0, 60)
plt.show()
```

### Interpretación 

Al analizar la gráfica generada por nuestra simulación de Monte Carlo, observamos la distribución geométrica de los puntos en el espacio muestral $\Omega = [0, 60] \times [0, 60]$, cuya área total es $|\Omega| = 60 \times 60 = 3600$.

La simulación refleja con alta fidelidad el comportamiento del continuo stocástico. Al lanzar miles de pares de variables uniformes $(X, Y)$, la proporción de puntos rojos sobre el plano replica el cálculo integral de la medida de Lebesgue. La ligera diferencia entre el valor teórico ($0.3055$) y el simulado ($0.3014$) se debe al error de muestreo aleatorio estándar, el cual disminuye a medida que aumentamos el número de iteraciones ($N \to \infty$) bajo la Ley de los Grandes Números.

El enfoque geométrico demuestra que en espacios muestrales continuos no numerables, la probabilidad de un evento no se calcula sumando puntos aislados (ya que la probabilidad de llegar en un instante infinitesimal exacto es cero, $\mathbb{P}(X=x) = 0$), sino comparando la proporción del área medible de la condición de interés respecto al área total del espacio de estados.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función para simular un volado individual
# la probabilidad 'p' aquí es de obtener SOL, que es nuestra variable de interés
def volado(p):
    # random.random() nos avienta un decimal uniforme entre 0 y 1
    # Si cae abajo del parámetro, se cuenta como éxito ("sol")
    return "sol" if random.random() < p else "aguila"

# Aquí vamos a guardar el resultado final de cada una de las 500 personas
p_lista = []

# Iniciamos el bucle externo: simulamos a las 500 personas en total
for i in range(500):
    
    # Cada persona va a lanzar la moneda exactamente 1000 veces
    N = 1000  
    resultados = [] # Limpiamos la lista para la persona en turno
    
    # Bucle interno: la persona 'i' realiza sus 1000 tiros individuales
    for j in range(N):
        # Usamos p = 0.63 porque estamos probando con una moneda cargada
        resultado = volado(0.63) 
        resultados.append(resultado) # Guardamos cada tiro en la lista
        
    # Contamos de forma exacta cuántos soles acumuló esta persona en sus 1000 tiros
    n_soles = resultados.count("sol")
    
    # Calculamos su probabilidad frecuentista (frecuencia relativa muestral)
    p_soles = n_soles / N  
    
    # Metemos el promedio de esta persona a nuestra lista global de control
    p_lista.append(p_soles)

# Calculamos el promedio acumulado paso a paso entre las personas (1 hasta 500)
# Esto sirve para ver cómo se va estabilizando la curva conforme agregamos datos
promedio_acumulado_volado = np.cumsum(p_lista) / np.arange(1, 501)

# Bloque de Graficación para el Reporte 
plt.figure(figsize=(10, 4))

# Graficamos el valor teórico que configuramos en la función como una línea base
plt.axhline(y=0.63, color='red', linestyle='--', label='Probabilidad Teórica Esperada (0.63)')

# Graficamos nuestra curva empírica acumulada para ver el comportamiento frecuentista
plt.plot(promedio_acumulado_volado, color='indigo', label='Frecuencia relativa acumulada (Soles)')

# Ponemos títulos y etiquetas claras
plt.title(' Convergencia Frecuentista del Volado Cargado ')
plt.xlabel('Número de experimentos (Personas virtuales)')
plt.ylabel('Proporción acumulada de Soles')

# Acotamos el eje Y para que se note el "efecto de aplanado" de la curva
plt.ylim(0.55, 0.70) 
plt.legend()
plt.grid(True) # Cuadrícula de fondo para medir visualmente la estabilidad
plt.show()

# Imprimimos el output final con 4 decimales para checar qué tan cerca quedó de 0.63
print(f"Resultado final promediado de la simulación: {promedio_acumulado_volado[-1]:.4f}")
```

### Interpretación

La variable que calculamos para cada persona es la probabilidad frecuentista individual:

$$\text{p\_soles} = \frac{\text{n\_soles}}{N}$$

Al promediar los resultados de las 500 repeticiones (`p_soles_prom = sum(p) / 500`), el output final reflejado en la consola es de **0.6296**.

A nivel de un solo intento o de muestras pequeñas (como la lista de 10 tiros donde se obtuvieron 3 soles y 7 águilas), la variabilidad aleatoria domina el escenario. Sin embargo, al extender el experimento a un tamaño muestral masivo de $500 \times 1000 = 500,000$ volados simulados en total, las desviaciones se compensan mutuamente, convergiendo con una precisión milimétrica hacia el parámetro teórico implícito de la función ($\lim_{n \to \infty} f_n = 0.63$).

El enfoque frecuentista demuestra que la repetición masiva de simulaciones computacionales (Método de Monte Carlo) es un mecanismo totalmente válido para aproximar y descubrir la verdadera medida de probabilidad de un fenómeno complejo cuando no se cuenta con una deducción analítica a priori.

## Ejemplo Propio: Simulación de Inspección de Calidad 

Se analiza un lote de artículos donde cada uno tiene una probabilidad $p = 0.3$ de presentar una característica de interés (éxito). 

Si tomamos muestras pequeñas repetidamente, podemos contar el número de éxitos obtenidos y verificar cómo la media de éxitos observada experimentalmente se aproxima al valor teórico esperado conforme aumentamos el número de simulaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parámetros basados en lo visto en clase
p_teorico = 0.3
n_ensayos = 1 # Cada artículo es un ensayo de Bernoulli (como n=1 en la clase)
num_simulaciones = 500

# Generamos la muestra de datos tal como lo hacíamos con binom.rvs
muestras = binom.rvs(n=n_ensayos, p=p_teorico, size=num_simulaciones, random_state=42)

# Calculamos el promedio acumulado de éxitos (frecuencia relativa)
promedio_acumulado = np.cumsum(muestras) / np.arange(1, num_simulaciones + 1)

# Graficamos la convergencia para el reporte del Notebook
plt.figure(figsize=(10, 4))
plt.axhline(y=p_teorico, color='green', linestyle='--', label=f'Probabilidad Teórica ({p_teorico})')
plt.plot(promedio_acumulado, color='darkblue', label='Frecuencia relativa acumulada')
plt.title('Propio: Convergencia de la proporción de éxitos en la muestra')
plt.xlabel('Número de muestras simuladas')
plt.ylabel('Proporción acumulada')
plt.legend()
plt.grid(True)
plt.show()

print(f"Resultado final de la proporción simulada: {promedio_acumulado[-1]:.4f}") # .4f nos da los primeros 4 decimales
```

### Interpretación

Al evaluar el comportamiento de la curva de nuestra simulación, observamos que en las primeras muestras el promedio acumulado exhibe un comportamiento altamente volátil (llegando a registrar picos cercanos a $0.65$ o caídas drásticas). Esto se debe a la varianza inherente de los lotes pequeños. Los momentos teóricos marginales para cada póliza son:
* **Esperanza Teórica ($\mathbb{E}[X_i]$):** $p = 0.3$
* **Varianza Teórica ($\text{Var}(X_i)$):** $p(1 - p) = 0.3 \times 0.7 = 0.21$

Conforme el número de carteras simuladas avanza hacia las 500 observaciones, la curva se estabiliza de manera asintótica sobre la franja verde discontinua, arrojando un valor empírico final de **0.3080**, el cual valida numéricamente la convergencia del estimador.

```python
# Simulación de probabilidad geométrica (Método de Montecarlo para el tiro al blanco)
n_dardos = 2000
x_dardos = np.random.uniform(-15, 15, n_dardos)
y_dardos = np.random.uniform(-15, 15, n_dardos)

# Verificar cuáles cayeron dentro del círculo de radio 15
dentro_circulo = (x_dardos**2 + y_dardos**2) <= 15**2
prob_geometrica_estimada = np.sum(dentro_circulo) / n_dardos
print(f"La probabilidad geométrica estimada es: {prob_geometrica_estimada:.4f}")
```

```python
# Simulación de Probabilidad Condicional

import pandas as pd

# 1. Reutilizamos el espacio muestral de los dos dados
dados_teoricos = [(i, j) for i in range(1, 7) for j in range(1, 7)]
df_condicional = pd.DataFrame(dados_teoricos, columns=['Dado 1', 'Dado 2'])

# 2. Definimos las máscaras booleanas para los eventos
# Evento A: Obtener el mismo resultado (Pares iguales)
df_condicional['Evento_A'] = df_condicional['Dado 1'] == df_condicional['Dado 2']

# Evento B: Que el primer dado sea par (2, 4, 6)
df_condicional['Evento_B'] = df_condicional['Dado 1'].isin([2, 4, 6])

# Evento A e Intersección B (A ∩ B): Son iguales Y el primero es par
df_condicional['Interseccion_A_B'] = df_condicional['Evento_A'] & df_condicional['Evento_B']

# 3. Cálculo de probabilidades clásicas usando cardinalidades
total_espacio = len(df_condicional)
p_b = len(df_condicional[df_condicional['Evento_B']]) / total_espacio
p_interseccion = len(df_condicional[df_condicional['Interseccion_A_B']]) / total_espacio

# Aplicamos la fórmula: P(A|B) = P(A ∩ B) / P(B)
p_a_dado_b = p_interseccion / p_b


print("             RESULTADOS DE PROBABILIDAD CONDICIONAL               ")
print(f"Probabilidad simple P(A) [Dados iguales]:               {len(df_condicional[df_condicional['Evento_A']])/total_espacio:.4f} (1/6)")
print(f"Probabilidad del evento condicionante P(B) [Dado 1 par]: {p_b:.4f} (1/2)")
print(f"Probabilidad de la intersección P(A ∩ B):                {p_interseccion:.4f} (3/36)")
print(f"PROBABILIDAD CONDICIONAL P(A | B):             {p_a_dado_b:.4f}")
```

```python
#Ejemplo independencia de eventos
# Definimos el espacio muestral del dado
dado = np.array([1, 2, 3, 4, 5, 6])
total_casos = len(dado)

# Definimos los eventos
evento_A = (dado % 2 == 0)       # Par
evento_B = (dado >= 5)           # Mayor o igual a 5
interseccion = evento_A & evento_B  # Par Y mayor o igual a 5 (es decir, el 6)

# Calculamos probabilidades
p_A = np.sum(evento_A) / total_casos
p_B = np.sum(evento_B) / total_casos
p_interseccion = np.sum(interseccion) / total_casos
p_producto = p_A * p_B

print(f" P(A): {p_A:.4f}")
print(f" P(B): {p_B:.4f}")
print(f" P(A ∩ B): {p_interseccion:.4f} (1/6)")
print(f" P(A)*P(B): {p_producto:.4f} (0.5 * 0.3333)")

# Comprobación de independencia
if np.isclose(p_interseccion, p_producto):
    print("\nLos eventos A y B son INDEPENDIENTES")
else:
    print("\nLos eventos A y B son DEPENDIENTES.")
```
