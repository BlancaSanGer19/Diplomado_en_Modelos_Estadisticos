## Probabilidad Laplaciana

Pra un experimento alelatorio, se tienen que $\Omega$ es el espacio muestral ( espacio de posibles resultados).

Sea $A$ un evento, entonces definimos la probabilidad de $A $ como sigue: 

$$ \mathbb {P}(A) = \frac{|A|}{|\Omega|}. $$

**Ejemplo.** Sea el experimento de lanzar una moneda 3 veces consecutivas, de manera independiente, es decir, el $i$-ésimo lanzamineto no afecta al $i+1$ lanzamiento. Es espacio muetsral es: 

$$ \Omega_{\text{moneda}} = \{ a,s\}$$

$$\Omega = \Omega_{\text{moneda}} \times \Omega_{\text{moneda}} \times \Omega_{\text{moneda}} = \{(s, s, s), (s, a, s), (s, s, a), (s, a, a), (a, s, a), (a, s, s), (a, a, s), (a, a, a)\}$$
y se tiene que $|\Omega|= 2^3 = 8$.

Sean los eventos:

- $A$: todos los resultados son águila 
- $S$: todos los resultados son sol
- $B$: por lo menos un resultado es sol (1 o 2 o 3 soles)

Vamos a calcular las probabilidades de $A, S y E$. Para esto, sea

- $A_i$ : cae águila en el $i$-ésimo lanzamiento. 

Para el evento $A$, se tienen que $A = A_1 \cap A_2 \cap A_3$




## 1. Primero vamos a crear el espacio muestral

```python
#Definimos el espacio muestral 
# Definimos el espacio muestral
Omega = set(product({'A', 'S'}, repeat=3)) # product genera tuplas de longitud 3 y set lo convierte en conjunto
Omega
```

## 2. Ahora vamos a crear el evento A 



```python
# Para el evento A_1: el primer elemento es águila
# Para el evento A_1: el primer elemento es águila
A_1 = { om for om in Omega if om[0] == 'A' }
A_1
```

## Probabilidad Condicional

Definimos la probabilidad condicional de que ocurra un evento $A$ **sabiendo** que ocurre un evento $B$ de:

$$\mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)} = \frac{|A \cap B|}{|B|}$$,

donde donde $\mathbb{P}(B) > 0$.

```python
#definimos la funcion de probabilidad condicional 
def prob_cond(E, F):
    return len(E & F) / len(F)
```

Queremos calcular la probabilidad condicional de los eventos: 

- $B$: el primer lanzamiento es águila
- $C$: dos de los resultados son águila

Queremos calcular $\mathbb{P}(B\mid C)$.

```python
#Para el evento B
B = {om for om in Omega if om[0] == 'A'}
B
```

```python
# Para el evento C:
C = { om for om in Omega if om.count('A')==2 }
C
```

```python
prob_cond(B,C)
```

```python
prob_cond(C,B)
```

## Independencia de eventos 

Decimos que dos eventos son independintes si 
$$ \mathbb{P}(A \cap B)= \mathbb{P}(A)\mathbb{P}(B)$$

```python
def prob(E):
    return len(E) / len(Omega)
```

```python
#Definimos la función de probabilidad condicional 
def indep(E,F):
    return prob(E & F ) == prob(E)*prob(F)
```

```python
indep(B,C)
```

### Ejemplo 

Sea el experimento aleatorio de lanzar un dado honesto dos veces. 
Sabemos que:  $$ \Omega = \{ (i,j) : i,j\in\{1,2,3,4,5,6\} \}$$
y además $ |\Omega|= 6^2 = 36 $

```python
# Vamos a crear el espacio muestral
Omega_dado = set(product([1,2,3,4,5,6], repeat=2))
#Omega_dado
```

```python
card_dado = len (Omega_dado)
card_dado
```

Definimos el siguinete evento: 

- $S_n$ : la suma de los dados es $n$, donde $n\in\{2,\dots,12\}$
  $$S_n = \{(i,j)\in\Omega \mid i+j =n \}$$

```python
# Evento S_n , un número anterior
def S(n):
    Sn = {(i,j) for i in range(1,7) for j in range(1,7) if i+j == n}
    # range genera una secuencias de numeros enteros de 1 al 6 
    return Sn
```

```python
S_siete = S(7)
S_siete
```
