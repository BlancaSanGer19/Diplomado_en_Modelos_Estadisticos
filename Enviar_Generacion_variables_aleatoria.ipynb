## 2. Métodos de generación de números aleatorios rectangulares

La generación de números con distribución uniforme $(0, 1)$ es el bloque fundamental, ya que cualquier otra distribución se construye a partir de estos números base.

## Características requeridas:
1. **Independencia:** Cada valor debe ser independiente de los anteriores.
2. **Equidistribución:** Cada posible valor en el rango $[a, b]$ debe tener la misma probabilidad $1/n$.
3. **Reproducibilidad:** La secuencia debe poder repetirse para propósitos de depuración o validación.
4. **Eficiencia:** Debe consumir poca memoria y ser computacionalmente rápida.

## 3. Métodos históricos:
* **Métodos manuales:** Lanzamiento de dados, sorteos físicos.
* **Tablas de biblioteca:** Libros con secuencias de números aleatorios precalculados.
* **Computación analógica:** Circuitos electrónicos que generan ruido.
* **Computación digital (Pseudoaleatoriedad):** Algoritmos deterministas que producen secuencias que *parecen* aleatorias (Ej: Congruenciales lineales).
La generación de números con distribución uniforme $(0, 1)$ es el bloque fundamental, ya que cualquier otra distribución se construye a partir de estos números base.


### Generación pseudoaleatoria

La generación de los números aleatorios rectangulares debe realizarse a través de relaciones matemáticas de recurrencia. Por esta razón se consideran **pseudoaleatorios**, ya que el proceso para generarlos es completamente determinístico.

Hay dos métodos que son los más utilizados para la generación. Ambos se basan en la siguiente definición:

> **Definición.** Dos enteros $a$ y $b$ son congruentes módulo $m$ si su diferencia es un múltiplo entero de $m$ y se expresa como:
> $$a \equiv b \pmod m$$

Como consecuencia de esto se tiene que:
* $(a - b)$ es divisible entre $m$.
* $a$ y $b$ dan el mismo residuo al ser divididos entre $m$.


### 4. Método congruencial multiplicativo

Este método permite generar una secuencia de números pseudoaleatorios uniformes en el intervalo $(0, 1)$ usando la siguiente fórmula recursiva:

$$n_{i+1} = a n_i \pmod m$$

Para garantizar las propiedades estadísticas óptimas y que el generador sea eficiente, se deben cumplir las siguientes condiciones de diseño:
* **$m$ (Módulo):** Debe ser tan grande como sea posible, dependiendo de los bits por palabra que maneje la computadora, descontando el bit del signo ($b$). Por lo tanto: $m = 2^b$.
* **$a$ (Multiplicador):** Debe satisfacer que $a \approx 2^{(b+1)/2}$ y que $a \equiv \pm 3 \pmod 8$. La segunda expresión equivale a que $(a - (\pm 3))$ es múltiplo de 8.
* **$n_0$ (Semilla):** Debe ser un entero positivo impar menor a $m$.
* **Periodo:** Bajo estas condiciones, el periodo máximo de números generados antes de que la secuencia se empiece a repetir será de longitud: $m/4$.

Para transformar los enteros generados $n_i$ en números pseudoaleatorios uniformes $u_i$ dentro del intervalo $(0,1)$, aplicamos la normalización:
$$u_i = \frac{n_i}{m}$$

### Normalización
Para obtener valores $u_i$ uniformes en el intervalo abierto $(0, 1)$, se realiza la transformación:
$$u_i = \frac{n_i}{m}$$

### Consideración Técnica
Aunque es eficiente computacionalmente, este método es **determinista**. Esto significa que si se conoce la semilla ($n_0$) y los parámetros ($a, m$), la secuencia completa puede ser replicada, lo cual es útil para fines de depuración y reproducibilidad en simulaciones, pero debe evitarse en aplicaciones que requieran seguridad criptográfica.

```python
# n0 : valor inicial
# a : multiplicador 
# m: módulo (2^b)
# n: número que quiero generar

def multiplicativo(n0,a,m,n):
    secuencia =[]  # lista vacia para ir gurdando los números 
    ni =n0 # iniciamos con un valor inicial
    # Repstimos n veces 
    for i in range(n):
        ni = (a*ni) % m # fórmula recurrente 
        secuencia.append(ni/m) #agreo el número a  lista normalizado (numeros entre (0,1))
    return secuencia    
```

```python
# Ejemplo:
m = 2**31
a = 65539 # a congruente +- 3(mod 8)
n0 = 123456789
n= 100

secuencia_mult = multiplicativo(n0,a,m,n)
print(secuencia_mult)
```

```python
# Ejmplo 2
def generador_congruencial_multiplicativo(semilla, a, m, iteraciones): # Genera una secuencia de números pseudoaleatorios uniformes U(0,1)
    # Verificación inicial de la semilla (debe ser impar)
    if semilla % 2 == 0:
        print("Advertencia: Se recomienda una semilla impar para maximizar el periodo.")
        
    secuencia_enteros = []
    secuencia_uniforme = []
    
    n_actual = semilla
    for _ in range(iteraciones):
        # Aplicamos la relación recursiva de congruencia
        n_siguiente = (a * n_actual) % m
        
        # Normalizamos para transformar al intervalo (0, 1)
        u = n_siguiente / m
        
        secuencia_enteros.append(n_siguiente)
        secuencia_uniforme.append(u)
        
        # Actualizamos para la siguiente iteración
        n_actual = n_siguiente
        
    return secuencia_enteros, secuencia_uniforme
```

### 5. Método congruencial mixto

A diferencia del método multiplicativo, el **Método Congruencial Mixto** incorpora una constante aditiva $c$ dentro de la relación de recurrencia. Esto le permite romper ciertas limitaciones de **periodo** y alcanzar una cobertura completa del módulo $m$.

La ecuación recursiva fundamental es:
$$n_{i+1} = (a n_i + c) \pmod m$$

Donde los parámetros de diseño deben cumplir con restricciones estrictas para asegurar que el generador sea de **periodo completo** (es decir, que genere exactamente $m$ números diferentes antes de repetirse):

* **$m$ (Módulo):** Determinado por la arquitectura de la computadora, $m = 2^b$.
* **$a$ (Multiplicador):** Debe satisfacer que $a \approx 2^{(b-1)/2}$ y la condición crucial $a \equiv 1 \pmod 4$, lo que significa que $(a - 1)$ es un múltiplo exacto de 4.
* **$c$ (Incremento):** Debe ser un entero positivo impar menor a $m$. Además, para un periodo completo, $c$ debe ser relativamente primo con $m$ (es decir, el máximo común divisor entre ambos debe ser 1).
* **$n_0$ (Semilla):** Un entero positivo impar menor a $m$.
* **Periodo:** Si se respetan las condiciones anteriores (respaldadas matemáticamente por el *Teorema de Hull-Dobell*), la longitud del periodo será exactamente **$m$**.

## Comparativa Matemática de Periodos 

| Propiedad / Método | Generador Multiplicativo | Generador Mixto |
| :--- | :--- | :--- |
| **Fórmula** | $n_{i+1} = a n_i \pmod m$ | $n_{i+1} = (a n_i + c) \pmod m$ |
| **Longitud Máxima de Periodo** | $\frac{m}{4}$ | $m$ (Periodo Completo) |
| **Restricción del multiplicador ($a$)** | $a \equiv \pm 3 \pmod 8$ | $a \equiv 1 \pmod 4$ |
| **Uso de Constante $c$** | No ($c = 0$) | Sí ($c > 0$ e impar) |

El método mixto es ampliamente preferido sobre el multiplicativo puro cuando los recursos de memoria o bits son limitados, dado que optimiza el uso de la arquitectura del procesador obteniendo cuatro veces más números únicos en su ciclo completo.

```python
# n0 : valor inicial
# a : multiplicador 
# m: módulo (2^b)
# n: número que quiero generar

def mixto(n0,a,m,n):
    secuencia =[]  # lista vacia para ir gurdando los números 
    ni =n0 # iniciamos con un valor inicial
    # Repstimos n veces 
    for i in range(n):
        ni = (a*ni+ c) % m # fórmula recurrente 
        secuencia.append(ni/m) #agreo el número a  lista normalizado (numeros entre (0,1))
    return secuencia   
```

```python
# Ejemplo:
m = 2**31
a = 1103515245 # a congruente 1 (mod 4)
c = 12345
n0 = 41
n= 100

secuencia_mixto = mixto(n0,a,m,n)
print(secuencia_mult)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generador_mixto(n0, a, c, m, cantidad):
    secuencia = []
    n = n0
    for _ in range(cantidad):
        n = (a * n + c) % m
        secuencia.append(n / m) # Normalización en (0, 1)
    return secuencia

# Parámetros de ejemplo (deben cumplir las condiciones del Teorema de Hull-Dobell)
m = 2**8    # Módulo
a = 17      # Multiplicador (a = 1 mod 4: 17 = 4*4 + 1)
c = 43      # Incremento (impar y primo relativo con m)
n0 = 5      # Semilla

# Generar 1000 números
datos = generador_mixto(n0, a, c, m, 1000)

# Visualización
plt.figure(figsize=(9, 5))
plt.hist(datos, bins=20, color='mediumseagreen', edgecolor='black', alpha=0.7)
plt.title('Distribución de números generados (Método Mixto)')
plt.xlabel('Intervalo (0, 1)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.show()
```

| **Método**     | **Fórmula**                            | **Ventajas**                                          | **Riesgos / Limitaciones**                                |
|----------------|----------------------------------------|-------------------------------------------------------|------------------------------------------------------------|
| Multiplicativo | $n_{i+1} = a n_i \mod m $           | Rápido, menos memoria                                 | Menor aleatoriedad, requiere $(a, m, n_0)$ bien elegidos |
| Mixto          | $ n_{i+1} = (a n_i + c) \mod m $     | Mejor distribución, periodo completo posible          | Más complejo, pero más robusto                             |
| NumPy (`uniform`) | Motor moderno y validado           | Alta calidad, probado, rápido                         | Caja negra, sin control del generador interno              |


### El Teorema de Hull-Dobell
La característica distintiva del Método Mixto frente al Multiplicativo es su capacidad de alcanzar un **periodo completo** (longitud $m$) sin necesidad de realizar cambios en la estructura de la arquitectura de la computadora. La garantía matemática de que el generador producirá exactamente $m$ números únicos antes de repetir la secuencia se formaliza a través del **Teorema de Hull-Dobell**, el cual establece que un generador congruencial lineal alcanza un periodo completo si y solo si:

1. **Relación entre $c$ y $m$:** La constante aditiva $c$ y el módulo $m$ son relativamente primos, es decir, $\text{mcd}(c, m) = 1$.
2. **Restricción sobre $a$:** Para cada número primo $p$ que divida a $m$, el valor $(a - 1)$ es divisible por $p$.
3. **Múltiplos de 4:** Si $m$ es divisible por 4, entonces $(a - 1)$ también debe ser divisible por 4.

La superioridad del método mixto sobre el multiplicativo puro ($c=0$) se fundamenta en la eliminación de la restricción del periodo máximo de $m/4$. Al incorporar la constante $c$, el generador rompe la dependencia de la semilla $n_0$ con respecto a la paridad y los ciclos de los subgrupos del módulo, logrando una cobertura exhaustiva del espacio de estados. 

* **Comportamiento en Espacios Limitados:** En sistemas donde la capacidad de representación de bits es restringida, la optimización que ofrece el método mixto es crítica, pues permite maximizar la entropía de la secuencia generada sin aumentar la complejidad computacional.
* **Implicaciones Estadísticas:** Aunque el método es lineal, la propiedad de "periodo completo" asegura que cada número entero en el conjunto $\{0, 1, \dots, m-1\}$ aparezca exactamente una vez por ciclo, lo cual minimiza el sesgo de selección en muestras finitas y maximiza la uniformidad de la distribución resultante tras la normalización.

### Consideraciones sobre la Aleatoriedad
Es imperativo recalcar que la "aleatoriedad" en este contexto es **pseudoaleatoriedad**. La secuencia es enteramente predecible si se conocen los parámetros iniciales $(n_0, a, c, m)$. Por lo tanto, la validez del método no descansa en la incognoscibilidad de los valores, sino en que la secuencia pase satisfactoriamente las pruebas de uniformidad (como la prueba de Kolmogórov-Smirnov) y de independencia (como la prueba de rachas), cumpliendo con los requisitos de las simulaciones de Monte Carlo.

## 5.Metodos de generación de numeros aleatorios no rectangulares

#### Método de la transformada inversa

Este método es la técnica estándar para generar variables aleatorias a partir de una distribución específica $f(x)$, utilizando como base una variable aleatoria uniforme $U \sim \text{Uniforme}(0, 1)$ generada previamente por los métodos congruenciales.

### Fundamento Teórico
Sea $X$ una variable aleatoria con función de distribución acumulada $F(x)$. Si $U$ es una variable aleatoria uniformemente distribuida en el intervalo $(0, 1)$, entonces la variable aleatoria $X$ definida por:
$$X = F^{-1}(U)$$
tiene exactamente la función de distribución $F(x)$.

### Pasos para la Aplicación
Para simular una variable aleatoria mediante este método, se debe seguir este algoritmo:

1. **Obtener la función de densidad:** Identificar la función de probabilidad o densidad $f(x)$ de la distribución deseada.
2. **Calcular la acumulada:** El método utiliza la función de distribución $F(x)$ de la distribución que se va a simular 
$$F(x) = \int_{-\infty}^{x} f(t)\,dt$$
3. **Invertir la función:** Despejar $x$ en la ecuación $u = F(x)$ para obtener la función inversa $x = F^{-1}(u)$.
4. **Generar uniformes:** Generar un número aleatorio $u$ en el intervalo $(0, 1)$ mediante un generador congruencial.
5. **Transformar:** Calcular el valor $x = F^{-1}(u)$, el cual será una observación aleatoria proveniente de la distribución objetivo.

### Consideraciones Críticas
* **Condición de invertibilidad:** El método requiere que $F(x)$ sea estrictamente creciente y continua, lo que garantiza la existencia de una función inversa $F^{-1}$ única.
* **Limitaciones:** En muchas distribuciones complejas, la integral de la densidad $f(x)$ no posee una forma cerrada o analítica (por ejemplo, la **Distribución Normal**), lo que hace imposible despejar $F^{-1}(u)$ algebraicamente. En tales casos, se debe recurrir a otros métodos como el de **Rechazo (Acceptance-Rejection)** o aproximaciones numéricas.

### Ventaja Comparativa
A diferencia de otros métodos, la Transformada Inversa preserva la estructura de la muestra (si se utiliza una secuencia de números uniformes con ciertas propiedades, esas propiedades se mapean directamente a la nueva distribución), siendo extremadamente eficiente cuando la inversa $F^{-1}$ puede expresarse de forma simple.



Como se sabe, los valores de $F(x)$ están en el intervalo $(0,1)$ al igual que los números rectangulares $U$. 

El método genera un $U$ y trata de determinar el valor de la variable aleatoria para la cual $F(x)$ sea igual a $U$.

Si $U \in (0,1)$:
$$ F(x) = U \quad \Rightarrow \quad x = F^{-1}(U)$$

$\mathbf{Teorema}.$ Sea $X$ una variable aleatoria real. Supongamos que su función de distribución $F$ es estrictamente creciente (por lo que $F$ es una biyección de $\mathbb{R}$ sobre $(0,1)$ y podemos denotar por $F^{-1}$ a su inversa). Sea $U\sim \text{unif}[0,1]$. Entonces $F^{-1}(U)$ tiene la misma ley que $X$.

Si $F$ no es estrictamente creciente, hemos visto que el teorema precedente sigue siendo válido bajo la condición de definir 
$$F^{-1}(u)=\inf\{ x\in \mathbb{R} : F(x)\geq u\},$$
la inversa generalizada de $F$.

```python
# Vamoa a genrar una v.a Bernoulli, usando una uniforme 
# Sabemos que x\sim Ber(p), si f(x)=p^{x} * (1-p)^{1-x}
import random #numeros aleatorios
def bernoulli(p):
    U = random.random() # U\sim Unif[0,1]
    if  U <= p:
        return 1 # éxito
    else:
        return 0 # fracaso
```

```python
# Simulamos n valores con p la probabilidad de éxito
p = 0.3
n = 10000
muestra =[bernoulli(p) for _ in range(n)]
print(muestra)
```

## Ejemplo: Distribución exponencial

Sea $\lambda>0$, entonces a través del teorema se puede generar una v.a. exponencial.

Sea $U \sim Unif[0,1]$, entonces si $X\sim Exp(\lambda)$, entonces

$$f_{X}(x) = \lambda e^{-\lambda x}$$
y que 
$$F_X(x) = 1 - e^{-\lambda x}$$

Sabemos que $1-e^{-\lambda x} = U$, entonces $1-U = e^{-\lambda x}$, y tomando logaritmo, se tiene que 
$ \ln(1-U) = -\lambda x$ y como $U$ es uniforme entonces $U \sim 1-U$, por lo que depejando a $x$, se tiene que 
$$ X = -\frac{\ln(U)}{\lambda} \sim Exp(\lambda) $$


```python
import numpy as np
import plotly.graph_objects as go # con un mayor control 
# Fijamos una semilla para reproducibilidad
np.random.seed(123)

# Damos el parámetro de la exponencial
lambd = 4

# Número de simulaciones
n = 10000

# Generamos n uniformes: U \sim Unif(0,1)
uniformes = np.random.random(n)

# Aplicamos la transformada inversa
# X = -ln(U) / lambd
exponenciales = -np.log(uniformes) / lambd

# Vamos a crear la curva de densidad teórica (exponencial)
# crea 300 valores igualmente espaciados entre 0 y el max(exponenciales)
x_vals = np.linspace(0, exponenciales.max(),300) 
y_vals = lambd * np.exp(-lambd * x_vals) # densidad de la exponencial

fig = go.Figure()
# Se van agregando capas a la figura

# Agregamos el histograma de la simulacion
fig.add_trace(go.Histogram(
    x = exponenciales, 
    nbinsx = 40, 
    histnorm = 'probability density',
    marker_color = 'skyblue',
    name = 'Simulación'
))

# Agregamos la curva de densidad teórica
fig.add_trace(go.Scatter(
    x = x_vals,
    y = y_vals, 
    mode = 'lines',
    name = 'Densidad teórica',
    line = dict(color='darkblue')
))

fig.update_layout(
    title = "Distribución exponencial simulada vs teórica (lambda = 4)",
    xaxis_title = 'x',
    yaxis_title = 'Densidad',
    bargap = 0.05 # espacio entre las barras
)

fig.show()

```

## Ejemplo: Distribución Uniforme Continua
$$f(x) = \frac{1}{b-a}, \quad a \leq x \leq b$$
$$F(x) = \frac{x-a}{b-a} = U \Rightarrow X = a + (b-a)U \sim Unif(a,b)$$

```python
import numpy as np
import plotly.graph_objects as go

# Fijamos una semilla para que los resultados sean reproducibles
np.random.seed(123)

#  Definimos los parámetros del intervalo [a, b] para la Uniforme Continua
a_param = 5
b_param = 15

# Número de simulaciones (usamos 10,000 para que se vea genial)
n = 10000

# Generamos n números uniformes estándar en el intervalo (0, 1)
U = np.random.random(n)

#  Aplicamos la ecuación de la transformación inversa derivada en clase:
# F(x) = (x - a) / (b - a) = U  =>  X = a + (b - a) * U
uniformes_continuas = a_param + (b_param - a_param) * U

#  Creamos la función de densidad teórica f(x) = 1 / (b - a)
x_vals = np.linspace(a_param - 2, b_param + 2, 500)
y_vals = np.where((x_vals >= a_param) & (x_vals <= b_param), 1 / (b_param - a_param), 0)

fig = go.Figure()

# Agregamos el histograma detallado con bastantes barras (nbinsx = 40)
fig.add_trace(go.Histogram(
    x = uniformes_continuas,
    nbinsx = 40,
    histnorm = 'probability density',
    marker_color = 'skyblue',
    name = 'Simulación'
))

# Agregamos la línea de la densidad teórica en color azul oscuro
fig.add_trace(go.Scatter(
    x = x_vals,
    y = y_vals,
    mode = 'lines',
    name = 'Densidad teórica',
    line = dict(color='darkblue', width=3)
))

# Configuración y estética de la gráfica interactiva
fig.update_layout(
    title = "Distribución Uniforme Continua Simulada vs Teórica Unif(5, 15)",
    xaxis_title = 'x',
    yaxis_title = 'Densidad',
    bargap = 0.05
)

fig.show()
```

## $\mathbf{Ejemplo}$ Distribució Cauchy 

Simulemos la variable aleatoria de Cauchy de parámetro $1$ que tiene por función de densidad 
$$f_{X}(x) = \frac{1}{\pi}\frac{1}{1+x^{2}}, \quad x\in\mathbb{R}$$

1. Corroborar que la función así definida es una función de densidad de probabilidad

  $$ \int_\mathbb{R}f_X(x)dx = 1$$

2. Encontrar $F_X = P(X\le x) = \int_{-\infty}^{x}f_{X}(t)dt$
3. Crear una gráfica en Python que muestre su forma
4.  Generar v.a. Cauchy a partir de una uniforme

Por el teorema anterior, para $u\in (0,1)$
$$u=\frac{1}{\pi}\arctan(x)+\frac{1}{2} \qquad \text{si y sólo si} \qquad x=\tan\left({\pi}\left(u-\frac{1}{2} \right) \right).$$

## Solución

### 1. Corroborar la validez de la función de densidad

A partir de la función de densidad de probabilidad para una variable aleatoria de Cauchy con parámetro de escala $1$:
$$f_X(x) = \frac{1}{\pi} \frac{1}{1 + x^2}, \quad x \in \mathbb{R}$$

Desarrollamos los puntos clave utilizando el **Teorema de la Transformación Inversa**:

Para que $f_X(x)$ sea una función de densidad legítima, su integral sobre todo el espacio de soporte $\mathbb{R}$ debe ser exactamente igual a $1$. Recordando que la derivada de $\arctan(x)$ es $\frac{1}{1+x^2}$:

$$\int_{-\infty}^{\infty} f_X(x)dx = \int_{-\infty}^{\infty} \frac{1}{\pi (1 + x^2)} dx = \frac{1}{\pi} \Big[ \arctan(x) \Big]_{-\infty}^{\infty}$$

Evaluando los límites al infinito ($\lim_{x \to \infty} \arctan(x) = \frac{\pi}{2}$ y $\lim_{x \to -\infty} \arctan(x) = -\frac{\pi}{2}$):

$$\frac{1}{\pi} \left( \frac{\pi}{2} - \left(-\frac{\pi}{2}\right) \right) = \frac{1}{\pi} (\pi) = 1$$



### 2. Encontrar la Función de Distribución Acumulada $F_X(x)$

La FDA se obtiene acumulando la probabilidad de la función desde $-\infty$ hasta un punto de corte $x$:
$$F_X(x) = \int_{-\infty}^{x} \frac{1}{\pi(1 + t^2)} dt = \frac{1}{\pi} \Big[ \arctan(t) \Big]_{-\infty}^{x}$$
$$F_X(x) = \frac{1}{\pi} \left( \arctan(x) - \left(-\frac{\pi}{2}\right) \right) = \frac{1}{\pi} \arctan(x) + \frac{1}{2}$$

La función de distribución acumulada (FDA) se obtiene integrando la densidad desde $-\infty$ hasta un punto de corte $x$:

$$F_X(x) = \mathbb{P}(X \le x) = \int_{-\infty}^{x} f_X(t)dt = \int_{-\infty}^{x} \frac{1}{\pi(1 + t^2)} dt$$

$$F_X(x) = \frac{1}{\pi} \Big[ \arctan(t) \Big]_{-\infty}^{x} = \frac{1}{\pi} \left( \arctan(x) - \left(-\frac{\pi}{2}\right) \right)$$

$$F_X(x) = \frac{1}{\pi} \arctan(x) + \frac{1}{2}$$



### 3. Crear una gráfica en Python que muestre su forma

```python
import numpy as np
import plotly.graph_objects as go

# Fijamos la semilla para reproducibilidad
np.random.seed(123)

# Número de simulaciones
n = 10000

# Generamos n uniformes estándar: U ~ Unif(0,1)
uniformes = np.random.random(n)

# Aplicamos la transformada inversa para la distribución Cauchy
# X = tan(pi * (U - 0.5))
cauchy_simuladas = np.tan(np.pi * (uniformes - 0.5))

# Nota matemática: Debido a las colas extremadamente pesadas de Cauchy,
# filtramos los valores muy extremos solo para que la gráfica se vea perfecta.
datos_filtrados = cauchy_simuladas[(cauchy_simuladas > -10) & (cauchy_simuladas < 10)]

# Creamos la curva de densidad teórica f(x) = 1 / (pi * (1 + x^2))
x_vals = np.linspace(-10, 10, 500)
y_vals = 1 / (np.pi * (1 + x_vals**2))

fig = go.Figure()

# Agregamos el histograma de la simulación con bastantes barras (40 barras)
fig.add_trace(go.Histogram(
    x = datos_filtrados,
    nbinsx = 40,
    histnorm = 'probability density',
    marker_color = 'skyblue',
    name = 'Simulación'
))

# Agregamos la curva de densidad teórica en azul oscuro
fig.add_trace(go.Scatter(
    x = x_vals,
    y = y_vals,
    mode = 'lines',
    name = 'Densidad teórica',
    line = dict(color='darkblue', width=3)
))

# Diseño general de la gráfica interactiva
fig.update_layout(
    title = "Distribución Cauchy Simulada vs Teórica (Parámetro = 1)",
    xaxis_title = 'x',
    yaxis_title = 'Densidad',
    bargap = 0.05,
    xaxis = dict(range=[-10, 10]) # Centramos la vista principal
)

fig.show()
```
