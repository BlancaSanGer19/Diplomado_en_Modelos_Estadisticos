# Probabilidad Laplaciana

Para un experimento aleatorio, se tiene que $\Omega$ es el espacio muestral (espacio de todos los resultados posibles).

La **Probabilidad Clásica o Laplaciana** es un enfoque axiomático que se utiliza bajo dos condiciones fundamentales:
1. El espacio muestral $\Omega$ debe ser **finito**.
2. Todos los resultados elementales deben ser **equiprobables** (es decir, tienen la misma posibilidad de ocurrir, como en una moneda o un dado honesto).

Si se cumplen estas condiciones y sea $A$ un evento (un subconjunto del espacio muestral, $A \subseteq \Omega$), definimos la probabilidad de ocurrencia de $A$ como la proporción entre los casos favorables y los casos totales: 

$$ \mathbb{P}(A) = \frac{\text{Casos Favorables}}{\text{Casos Totales}} = \frac{|A|}{|\Omega|}. $$


**Ejemplo Práctico.** Sea el experimento de lanzar una moneda 3 veces consecutivas de manera independiente; es decir, el resultado del $i$-ésimo lanzamiento no afecta ni probabilística ni físicamente al $i+1$ lanzamiento.

El espacio muestral elemental para un solo lanzamiento es $\Omega_{\text{moneda}} = \{ a, s\}$. Al realizar tres lanzamientos independientes, construimos el espacio muestral completo mediante el producto cartesiano: 

$$ \Omega = \Omega_{\text{moneda}} \times \Omega_{\text{moneda}} \times \Omega_{\text{moneda}} $$

$$ \Omega = \{(s, s, s), (s, a, s), (s, s, a), (s, a, a), (a, s, a), (a, s, s), (a, a, s), (a, a, a)\} $$

Aplicando el principio fundamental del conteo, se tiene que la cardinalidad total de nuestro espacio es $|\Omega|= 2^3 = 8$.

A partir de este espacio, nos interesa modelar y analizar el comportamiento de los siguientes eventos:
* **$A$**: Todos los resultados son águila. Es un evento simple compuesto únicamente por la tupla $(a, a, a)$.
  
* **$S$**: Todos los resultados son sol. Compuesto únicamente por la tupla $(s, s, s)$.
* **$B$**: Por lo menos un resultado es sol (puede caer 1, 2 o hasta 3 soles en total). Este es un evento compuesto que representa matemáticamente el complemento del evento $A$ ($B = A^c$).

Para facilitar la construcción computacional mediante intersecciones de conjuntos, definamos primero el evento base:
* $A_i$ : Cae águila en el $i$-ésimo lanzamiento ($i \in \{1, 2, 3\}$).

De este modo, podemos expresar formalmente el evento de obtener tres águilas consecutivas como la ocurrencia simultánea de cada uno de ellos:

$$ A = A_1 \cap A_2 \cap A_3 $$

# Probabilidad Condicional

La **probabilidad condicional** nos permite incorporar nueva información o "evidencia" al cálculo de probabilidades. En lugar de evaluar la probabilidad de que ocurra un evento $A$ sobre todo el espacio muestral original $\Omega$, calculamos la probabilidad de que ocurra $A$ **sabiendo** o dado que ya ha ocurrido (o es seguro que ocurrirá) un evento previo $B$.

Formalmente, si $\mathbb{P}(B) > 0$, la probabilidad condicional de $A$ dado $B$ se define como:

$$ \mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)} $$

### Interpretación y Enfoque Clásico (Laplaciano)
Desde la perspectiva clásica, al saber que el evento $B$ ha ocurrido, nuestro espacio de resultados posibles se **reduce** o se "actualiza", pasando de ser $\Omega$ a ser únicamente el conjunto $B$. Los únicos casos que ahora son favorables para $A$ son aquellos resultados que están simultáneamente en $A$ y en $B$ (es decir, la intersección $A \cap B$). Por lo tanto, la fórmula se puede simplificar mediante las cardinalidades de los conjuntos:

$$ \mathbb{P}(A \mid B) = \frac{\frac{|A \cap B|}{|\Omega|}}{\frac{|B|}{|\Omega|}} = \frac{|A \cap B|}{|B|} $$

### Propiedades Claves:
1. **Regla del Producto (Multiplicativa):** A partir de la definición, podemos despejar la probabilidad de la intersección. Esto es sumamente útil para calcular la ocurrencia simultánea de eventos secuenciales:
   $$ \mathbb{P}(A \cap B) = \mathbb{P}(B) \cdot \mathbb{P}(A \mid B) $$
2. **Conexión con el Teorema de Bayes:** La probabilidad condicional es la piedra angular del análisis bayesiano, permitiéndonos invertir el orden de la condicionalidad mediante la relación:
   $$ \mathbb{P}(B \mid A) = \frac{\mathbb{P}(A \mid B)\mathbb{P}(B)}{\mathbb{P}(A)} $$

```python
#definimos la funcion de probabilidad condicional 
def prob_cond(E, F):
    return len(E & F) / len(F)
```

# Independencia de Eventos 


En la teoría de la probabilidad, la **independencia** describe una situación en la que la ocurrencia o no ocurrencia de un evento no altera en lo absoluto la probabilidad de que ocurra el otro. 

Formalmente, dos eventos $A$ y $B$ son independientes si y solo si la probabilidad de su intersección (su ocurrencia simultánea) es igual al producto de sus probabilidades individuales:
$$ \mathbb{P}(A \cap B) = \mathbb{P}(A)\mathbb{P}(B) $$

### Interpretación mediante Probabilidad Condicional
La manera más intuitiva de comprender la independencia es conectándola con la probabilidad condicional. Si recordamos que:
$$ \mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)} $$

Y sustituimos la condición de independencia ($\mathbb{P}(A \cap B) = \mathbb{P}(A)\mathbb{P}(B)$), obtenemos que:
$$ \mathbb{P}(A \mid B) = \frac{\mathbb{P}(A)\mathbb{P}(B)}{\mathbb{P}(B)} = \mathbb{P}(A) $$

Esto significa que **saber que $B$ ya ocurrió no nos da absolutamente ninguna información nueva** para modificar nuestra predicción sobre $A$; la probabilidad de $A$ sigue siendo exactamente la misma. De igual manera, se cumple que $\mathbb{P}(B \mid A) = \mathbb{P}(B)$.

### Nota Importante: Independencia vs. Mutua Exclusión
Existe una confusión muy común entre estos dos conceptos:
*   **Eventos mutuamente excluyentes (o disjuntos):** Son eventos que **no pueden ocurrir al mismo tiempo** ($A \cap B = \emptyset$, por lo que $\mathbb{P}(A \cap B) = 0$). Si ocurre $A$, la probabilidad de que ocurra $B$ se vuelve automáticamente cero. Por lo tanto, ¡están fuertemente relacionados!
*   **Eventos independientes:** Son eventos que **sí pueden ocurrir al mismo tiempo**, pero la presencia de uno no afecta al otro. 

*Regla de oro:* Si dos eventos tienen probabilidad mayor a cero, **no pueden ser mutuamente excluyentes e independientes al mismo tiempo**.

```python
def prob(E):
    return len(E) / len(Omega)
```

```python
#Definimos la función de probabilidad condicional 
def indep(E,F):
    return prob(E & F ) == prob(E)*prob(F)
```

### Ejemplo 

Sea el experimento aleatorio de lanzar un dado honesto dos veces. 
Sabemos que:  $$ \Omega = \{ (i,j) : i,j\in\{1,2,3,4,5,6\} \}$$
y además $ |\Omega|= 6^2 = 36 $

```python
# Vamos a crear el espacio muestral
Omega_dado = set(product([1,2,3,4,5,6], repeat=2))
#Omega_dado
```

```python
card_dado = len (Omega_dado)
card_dado
```

Definimos el siguiente evento: 

* **$S_n$** : la suma de los dados es $n$, donde $n \in \{2, \dots, 12\}$.
  $$ S_n = \{(i,j) \in \Omega \mid i+j = n \} $$.

```python
# Función para generar el evento de que la suma de las caras sea n
def S(n):
    # range(1,7) genera los números enteros del 1 al 6 
    Sn = {(i, j) for i in range(1, 7) for j in range(1, 7) if i + j == n} 
    return Sn 
```

```python
# Obtenemos el evento para cuando la suma sea 7
S_siete = S(7)
S_siete 
```

```python
# Calculamos la probabilidad de que la suma de los dos dados sea 7
len(S_siete) / len(Omega_dado) 
```
